# ðŸ”¸ Formalism of Quantum Mechanics: Unitary time evolution

We begin by formulating our **second axiom** of quantum mechanics that tells us how a quantum system evolves in time. The formulation builds on the mathematical formalism that we have developed in the previous chapter. 


<strong>Axiom II (Quantum Dynamics):</strong><br><br>
  The time evolution of a closed quantum system is governed by a unitary transformation. If the system is in state $|\psi(t_0)\rangle$ at time $t_0$, its state at a later time $t$ is given by
  
$$
  |\psi(t)\rangle = U(t, t_0)|\psi(t_0)\rangle,
$$

  where $U(t, t_0)$ is a unitary operator on $\mathcal{H}$. 


### ðŸ”¹ Unitary Time Evolution: Essence and Physical Meaning

The statement of **Axiom II** â€” that the time evolution of a closed quantum system is governed by a **unitary operator** â€” has profound physical and mathematical consequences. We recall that an operator $U(t, t_0)$ on the Hilbert space $\mathcal{H}$ is **unitary** if:

$$
U(t, t_0)^\dagger U(t, t_0) = U(t, t_0) U(t, t_0)^\dagger = I
$$

where $U^\dagger$ denotes the **Hermitian adjoint** (conjugate transpose) of $U$, and $I$ is the identity operator. The unitary transformations generally:

- **Preserve inner products**:
  $$
  \langle \phi(t) | \psi(t) \rangle = \langle \phi(t_0) | \psi(t_0) \rangle
  $$
- **Preserve norms**:
  $$
  \| |\psi(t)\rangle \| = \| |\psi(t_0)\rangle \|
  $$
- **Preserve total probability**:
  $$
  \langle \psi(t) | \psi(t) \rangle = 1 \quad \text{if and only if} \quad \langle \psi(t_0) | \psi(t_0) \rangle = 1
  $$

The **essence** of unitary time evolution is the idea that:

> **In a closed system, the information contained in the quantum state is neither created nor destroyed as time passes.**

- There is no loss of probability (no decay or dissipation).
- The evolution is **reversible**: for every $U(t, t_0)$, the inverse $U(t_0, t) = U(t, t_0)^\dagger$ exists.
- The structure of quantum superpositions is maintained over time.

In other words, for a closed system **quantum dynamics is perfectly coherent** for all times.

### ðŸ”¹ Differential equation for the state vector: SchrÃ¶dinger equation

Having a unitary time-evolution at our disposal, we can calculate the state vector $| \psi (t) \rangle$ at any finite time $t$, given some initial time $t_0$. This gives us the global time-evolution of the system. If we would like to know how the system changes locally around some time, it would be useful to have a differential equation of motion for the state vector. We can derive such an equation starting from the fundamental postulate by differentiation with respect to time:

$$
| \psi(t) \rangle = {U}(t, t_0) | \psi(t_0) \rangle
$$

where $\hat{U}(t, t_0)$ is the unitary time-evolution operator. Taking the time derivative, we get:

$$
\frac{d}{dt} | \psi(t) \rangle = \frac{d}{dt} {U}(t, t_0) | \psi(t_0) \rangle = \left( \frac{\partial {U}(t, t_0) }{ \partial t}  \right) | \psi(t_0) \rangle
$$

To express this in terms of the state at time $t$, we recognize that $| \psi(t_0) \rangle = {U}^\dagger(t, t_0) | \psi(t) \rangle$, so that:

$$
\frac{d}{dt} | \psi(t) \rangle = \left( \frac{\partial {U}(t, t_0)}{\partial t}  {U}^\dagger(t, t_0) \right) | \psi(t) \rangle
$$

or equivalently,

$$
\frac{d}{dt} | \psi(t) \rangle = \left( \frac{\partial {U}(t, t_0)}{\partial t}  {U}(t_0, t) \right) | \psi(t) \rangle
$$



We define:

$$
\Omega(t) := \frac{\partial U(t,t_0)}{\partial t} \, U(t_0,t)
$$

Thus, the time derivative of the state becomes:

$$
\frac{d}{dt} |\psi(t)\rangle = \Omega(t) |\psi(t)\rangle
$$

We have now an equation of motion for the state vector whose structure is determined by the operator $\Omega(t)$ whose properties we study in the following section.

---

## ðŸ”¹ Properties of $\Omega(t)$

Since $U(t,t_0)$ is a **unitary operator**, it satisfies:

$$
U(t,t_0)^\dagger U(t,t_0) = I
$$

Differentiating both sides with respect to $t$, we obtain:

$$
\frac{\partial}{\partial t} \left( U(t,t_0)^\dagger U(t,t_0) \right) = 0
$$

Applying the product rule:

$$
\left( \frac{\partial U(t,t_0)^\dagger}{\partial t} \right) U(t,t_0) + U(t,t_0)^\dagger \left( \frac{\partial U(t,t_0)}{\partial t} \right) = 0
$$

Now, recall that:

$$
\frac{\partial U(t,t_0)^\dagger}{\partial t} = \left( \frac{\partial U(t,t_0)}{\partial t} \right)^\dagger
$$

because differentiation commutes with the Hermitian conjugate.  
Thus:

$$
\left( \frac{\partial U(t,t_0)}{\partial t} \right)^\dagger U(t,t_0) + U(t,t_0)^\dagger \left( \frac{\partial U(t,t_0)}{\partial t} \right) = 0
$$

Multiply this equation from the right by $U(t_0,t)$, using $U(t_0,t) = U(t,t_0)^\dagger$, and obtain:

$$
\left( \frac{\partial U(t,t_0)}{\partial t} U(t_0,t) \right)^\dagger + \left( \frac{\partial U(t,t_0)}{\partial t} U(t_0,t) \right) = 0
$$

Thus:

$$
\boxed{ \Omega(t)^\dagger + \Omega(t) = 0 }
$$

That is, $\Omega(t)$ is an **anti-Hermitian operator**.

---

## ðŸ”¹ Definition of the Hamiltonian

Having established that $\Omega(t)$ is anti-Hermitian, it is natural to define the a Hermitian operator called **Hamiltonian** $\hat{H}(t)$ by:

$$
\boxed{ \hat{H}(t) := i\hbar \, \Omega(t) }
$$

The factor $i\hbar$ ensures that $\hat{H}(t)$ is **Hermitian**.

Substituting back into the expression for the derivative of the state:

$$
\frac{d}{dt} |\psi(t)\rangle = \Omega(t) |\psi(t)\rangle = \frac{1}{i\hbar} \hat{H}(t) |\psi(t)\rangle
$$

thus:

$$
i\hbar \frac{d}{dt} |\psi(t)\rangle = \hat{H}(t) |\psi(t)\rangle
$$

We have now derived the **time-dependent SchrÃ¶dinger equation** from the unitary evolution structure.



### ðŸ”¹ Unitary Operators and Norm Conservation

In quantum mechanics, the evolution of a closed system must conserve probabilities. This is mathematically guaranteed if the evolution is governed by a **unitary operator**.

#### ðŸ”¸ Definition (Unitary Operator):

A linear operator $U$ on a Hilbert space $\mathcal{H}$ is called **unitary** if it preserves the norm of every state vector:

$$
\|U|\psi\rangle\| = \||\psi\rangle\| \quad \text{for all } |\psi\rangle \in \mathcal{H}.
$$

Equivalently, $U$ is unitary if it satisfies:

$$
U^\dagger U = U U^\dagger = \mathbb{I},
$$

where $U^\dagger$ is the Hermitian adjoint of $U$, and $\mathbb{I}$ is the identity operator.

---

### ðŸ”¹ Norm Conservation Implies Unitarity

Suppose $U$ is a linear operator satisfying:

$$
\langle U\psi | U\psi \rangle = \langle \psi | \psi \rangle \quad \text{for all } |\psi\rangle \in \mathcal{H}.
$$

Then:

$$
\langle \psi | U^\dagger U | \psi \rangle = \langle \psi | \psi \rangle.
$$

Subtracting both sides gives:

$$
\langle \psi | (U^\dagger U - \mathbb{I}) | \psi \rangle = 0 \quad \text{for all } |\psi\rangle.
$$

We now prove that this condition implies $U^\dagger U = \mathbb{I}$.

---

### ðŸ”¹ Lemma: Vanishing Quadratic Form Implies Zero Operator

Let $A$ be a Hermitian operator on a complex Hilbert space $\mathcal{H}$. Suppose:

$$
\langle \psi | A | \psi \rangle = 0 \quad \text{for all } |\psi\rangle \in \mathcal{H}.
$$

Then:

$$
A = 0.
$$

#### ðŸ”¸ Proof:

We apply the **polarization identity** in a complex Hilbert space:

$$
\langle \phi | A | \psi \rangle = \frac{1}{4} \sum_{k=0}^{3} i^k \langle \phi + i^k \psi | A | \phi + i^k \psi \rangle.
$$

Since each quadratic term on the right-hand side vanishes by assumption, it follows that:

$$
\langle \phi | A | \psi \rangle = 0 \quad \text{for all } |\phi\rangle, |\psi\rangle.
$$

Hence, all matrix elements of $A$ vanish and $A = 0$.

---

### ðŸ”¹ Conclusion

Applying the lemma to $A = U^\dagger U - \mathbb{I}$ (which is Hermitian), we conclude:

$$
U^\dagger U = \mathbb{I}.
$$

Similarly, norm preservation of $U^\dagger$ implies $U U^\dagger = \mathbb{I}$. Therefore, $U$ is **unitary**:

$$
U^\dagger = U^{-1}.
$$
