# ðŸ”¸ Linear maps, operators and functionals

A natural way to understand any mathematical structure is to study mappings
that preserve that structure. 
This idea is in the core of mathematics and also physics: we gain insight into a structure or a system not only by studying its **elements**, but by examining its **symmetries** and **transformations** that respect the rules of the structure.

In the case of **vector spaces**, the core structure is linearity: vector addition and scalar multiplication. So the natural class of mappings to study are those that **preserve linear sturcture**. Such maps must have two propersties: **additivity** and **homogeneity**. 

In the following we define such maps depending on what the domain and the target space is. Depending on that slightly different terminology is used.

---

### ðŸ”¹ Definition: Linear Map

Let $ V $ and $ W $ be vector spaces over the same field $\mathbb{F} $ (typically $ \mathbb{C} $).

A function$T: V \to W$ is called a **linear map** if it satisfies the following two conditions for all $ |\psi\rangle, |\phi\rangle \in V$ and all scalars $a \in \mathbb{F} $:

1. **Additivity**:

$$
 T(|\psi\rangle + |\phi\rangle) = T|\psi\rangle + T|\phi\rangle
$$

2. **Homogeneity**:

$$
T(a |\psi\rangle) = a \cdot T|\psi\rangle
$$

---

### ðŸ”¹ Definition: Linear Operator

A **linear operator** is a linear map from a vector space to itself.

Let $ V $ be a vector space over $\mathbb{F}$. A function $ A: V \to V $ is a **linear operator** if:

1. **Additivity**:

$$
A(|\psi\rangle + |\phi\rangle) = A|\psi\rangle + A|\phi\rangle
$$

2. **Homogeneity**:
 
$$
A(a |\psi\rangle) = a \cdot A|\psi\rangle
$$

---

### ðŸ”¹ Definition: Linear Functional

Let $V$ be a vector space over $ \mathbb{F}$. A **linear functional** is a linear map from $ V$ to the field $\mathbb{F}$:

$$
f: V \to \mathbb{F}
$$

It satisfies the same linearity conditions:

1. **Additivity**:

$$
f(|\psi\rangle + |\phi\rangle) = f(|\psi\rangle) + f(|\phi\rangle)
$$

2. **Homogeneity**:

$$
f(a |\psi\rangle) = a \cdot f(|\psi\rangle)
$$

Linear functionals are the elements of the **dual space** $V^*$, which we will define next.

In Dirac notation, linear functionals appear as **bras** as we will see later:

$$
\langle \psi | : V \to \mathbb{C}, \quad |\phi\rangle \mapsto \langle \psi | \phi \rangle
$$



## ðŸ”¸ The Riesz Representation Theorem (Finite-Dimensional Case)

Let $V$ be a finite-dimensional complex inner product space with inner product $\langle \cdot | \cdot \rangle$.

Then for **every linear functional** $f \in V^*$ (that is, every linear map $f: V \to \mathbb{C}$), there exists a **unique vector** $|\phi\rangle \in V$ such that:

$$
f(|\psi\rangle) = \langle \phi | \psi \rangle \quad \text{for all } |\psi\rangle \in V
$$

This is called the **Riesz representation** of the functional $f$ by the vector $|\phi\rangle$.

---

## ðŸ”¹ Proof (Finite-Dimensional Case)

Let $\dim V = n$ and let $\{ |e_i\rangle \}_{i=1}^n$ be an orthonormal basis of $V$.

Let $f \in V^*$ be a linear functional. We want to find a vector $|\phi\rangle \in V$ such that:

$$
f(|\psi\rangle) = \langle \phi | \psi \rangle \quad \forall |\psi\rangle \in V
$$

Write any $|\psi\rangle \in V$ as:

$$
|\psi\rangle = \sum_{i=1}^n \psi_i |e_i\rangle
$$

Since $f$ is linear:

$$
f(|\psi\rangle) = \sum_{i=1}^n \psi_i f(|e_i\rangle)
$$

Now define:

$$
|\phi\rangle := \sum_{i=1}^n f(|e_i\rangle)^* |e_i\rangle
$$

Then compute:

$$
\begin{aligned}
\langle \phi | \psi \rangle 
&= \left\langle \sum_{i=1}^n {f(|e_i\rangle)^* |e_i\rangle} \Big| \sum_{j=1}^n \psi_j |e_j\rangle \right\rangle \\
&= \sum_{i,j=1}^n f(|e_i\rangle)^* \psi_j \langle e_i | e_j \rangle \\
&= \sum_{i=1}^n f(|e_i\rangle)^* \psi_i \\
&= \sum_{i=1}^n \psi_i f(|e_i\rangle) = f(|\psi\rangle)
\end{aligned}
$$

Thus $f(|\psi\rangle) = \langle \phi | \psi \rangle$ for all $|\psi\rangle$, and the vector $|\phi\rangle$ is unique due to the non-degeneracy of the inner product, completing the proof.


---

## ðŸ”¹ Defining Bras Using Riesz

We now define the **dual vector** (bra) corresponding to a ket $|\psi\rangle \in V$ as the unique linear functional:

$$
\langle \psi | : V \to \mathbb{C}, \quad |\phi\rangle \mapsto \langle \psi | \phi \rangle
$$

By the Riesz theorem, this bra $\langle \psi |$ is an element of the **dual space** $V^*$.

This defines a **bijection**:

$$
V \xrightarrow{\text{Riesz}} V^*, \quad |\psi\rangle \mapsto \langle \psi |
$$

---

### ðŸ§  Interpretation

- The inner product $\langle \psi | \phi \rangle$ is interpreted as **applying the bra** $\langle \psi |$ to the ket $|\phi\rangle$.
- The Riesz theorem tells us that every linear functional on $V$ arises in this way.
- This justifies Diracâ€™s notation commonly used in physics:

$$
  \langle \psi | \phi \rangle = \text{â€œapply the functional } \langle \psi | \text{ to the vector } |\phi\rangle \text{"}
$$

## ðŸ”¸ Riesz Representation Theorem in Infinite Dimensions

In the finite-dimensional case, the Riesz Representation Theorem gives a **bijective correspondence** between a Hilbert space $V$ and its dual space $V^*$ via the inner product:

$$
f \in V^* \quad \Longleftrightarrow \quad \exists! \, |\phi\rangle \in V \text{ such that } f(|\psi\rangle) = \langle \phi | \psi \rangle \quad \forall |\psi\rangle \in V
$$

This result extends â€” with some technical care â€” to **infinite-dimensional Hilbert spaces**.

---

### ðŸ”¹ Infinite-Dimensional Version (Rieszâ€“FrÃ©chet Theorem)

Let $\mathcal{H}$ be a (possibly infinite-dimensional) **Hilbert space** over $\mathbb{C}$.

Then for every continuous linear functional $f : \mathcal{H} \to \mathbb{C}$, there exists a **unique** vector $|\phi\rangle \in \mathcal{H}$ such that:

$$
f(|\psi\rangle) = \langle \phi | \psi \rangle \quad \text{for all } |\psi\rangle \in \mathcal{H}
$$

> ðŸ§  The crucial difference is the requirement that $f$ be **continuous**. In infinite dimensions, not all linear functionals are continuous â€” but the Riesz representation holds **exactly** for those that are.

This general form justifies identifying the **continuous dual** $\mathcal{H}^*$ with $\mathcal{H}$ itself via the inner product.

## ðŸ”¸ From Dual Vectors to Adjoint Operators

Now that we understand bras as elements of the dual space $V^*$, and the inner product as a bridge between kets and bras via Riesz, we can move on to defining the **adjoint of a linear operator**.

---

## ðŸ”¹ Definition: Adjoint Operator

Let $A : \mathcal{H} \to \mathcal{H}$ be a linear operator on a Hilbert space $\mathcal{H}$.

The **adjoint operator** $A^\dagger : \mathcal{H} \to \mathcal{H}$ is the unique operator satisfying:

$$
\langle \phi | A |\psi\rangle = \langle A^\dagger \phi | \psi \rangle \quad \text{for all } |\psi\rangle, |\phi\rangle \in \mathcal{H}
$$

> In words: the **action of $A$ on the ket $|\psi\rangle$**, tested by the bra $\langle \phi |$, is **equivalent** to the **action of $A^\dagger$ on the bra**.

This relation defines $A^\dagger$ completely, due to the non-degeneracy of the inner product.




### ðŸ”¸ Theorem (Existence and Uniqueness of the Adjoint Operator)


Let $A: \mathcal{H} \to \mathcal{H}$ be a linear operator on a finite-dimensional Hilbert space $\mathcal{H}$.

Then:

âœ… There exists a **unique** linear operator $A^\dagger: \mathcal{H} \to \mathcal{H}$ such that:

$$
\langle \phi | A |\psi\rangle = \langle A^\dagger \phi | \psi \rangle \quad \text{for all } |\phi\rangle, |\psi\rangle \in \mathcal{H}.
$$

The operator $A^\dagger$ is called an **adjoint** of the operator $A$.

---

ðŸ”¹ Proof 

This theorem can be proven by using the Riesz theorem. 

- For each fixed $|\phi\rangle \in \mathcal{H}$, we define:

$$
\Phi_\phi : \mathcal{H} \to \mathbb{C}, \quad \Phi_\phi(|\psi\rangle) := \langle \phi | A |\psi\rangle.
$$

- $\Phi_\phi$ is linear a linear functional in $|\psi\rangle$ as can be proven by using the linearity of $A$ and the linearity of the scalar product in the second theorem.
- By the **Riesz Representation Theorem**, there exists a **unique** vector $A^\dagger |\phi\rangle \in \mathcal{H}$ such that:

$$
\Phi_\phi(|\psi\rangle) = \langle A^\dagger \phi | \psi \rangle \quad \text{for all } |\psi\rangle \in \mathcal{H}.
$$

Thus:

$$
\langle \phi | A |\psi\rangle = \langle A^\dagger \phi | \psi\rangle
$$

for all $|\phi\rangle, |\psi\rangle \in \mathcal{H}$.

The operator $A^\dagger$ is therefore **well-defined** and is also **uniquely determined**.

---


## ðŸ”¹ Self-Adjoint and Hermitian Operators

An operator $A$ is called **self-adjoint** if:

$$
A = A^\dagger.
$$

In physics, self-adjoint operators are often also referred to as **Hermitian operators**. As we already know, in quantum mechanics, **observables** (measurable quantities like position, momentum, energy) are represented by **self-adjoint operators** because they have a nice property that their eigenvalues are real.
One of the most important mathematical facts about self-adjoing operators is the **spectral theorem** that we now shortly address.

---



> **Spectral Theorem (finite-dimensional version):**  
> If $A$ is a self-adjoint operator on a finite-dimensional Hilbert space $\mathcal{H}$, then:
> - All eigenvalues of $A$ are real.
> - $\mathcal{H}$ has an orthonormal basis consisting of eigenvectors of $A$.
> - $A$ can be diagonalized as:
>   $$
>   A = \sum_k \lambda_k |\lambda_k\rangle \langle \lambda_k|,
>   $$
>   where $\lambda_k$ are the eigenvalues and $|\lambda_k\rangle$ the corresponding orthonormal eigenvectors.

Thus, self-adjoint operators behave analogously to symmetric matrices over $\mathbb{R}$: they can be fully diagonalized with an orthonormal basis.

---

## ðŸ”¹ Definition: Unitary Operators

A linear operator $U: \mathcal{H} \to \mathcal{H}$ is called **unitary** if:

$$
U^\dagger U = U U^\dagger = I.
$$

Equivalently, $U$ preserves the inner product:

$$
\langle U\phi | U\psi \rangle = \langle \phi | \psi \rangle \quad \text{for all } |\phi\rangle, |\psi\rangle \in \mathcal{H}.
$$
 
Uunitary operators generalize the idea of **rotations** to the Hilbert spaces: they shuffle vectors into other vectors **without changing their norms or inner products**.

---

### ðŸ§  Key Insights

- **Adjoint operators** generalize the idea of taking the conjugate transpose of matrices.
- **Self-adjoint (Hermitian) operators** represent **observable quantities** and have **real eigenvalues**.
- The **spectral theorem** ensures that any Hermitian operator can be understood via its eigenvectors and eigenvalues.
- **Unitary operators** represent **symmetries** and **time evolution** in quantum mechanics, preserving probabilities and inner products.
