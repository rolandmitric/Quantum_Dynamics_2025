
## A Historic View on the Heisenberg Formulation of Quantum Mechanics

When we teach quantum mechanics today, we usually begin with Schrödinger’s formulation, which is built on the wavefunction and the corresponding "wave equation." Because we are intuitively familiar with the concept of waves, we tend to accept Schrödinger’s “wave mechanics” as a natural starting point. This is one reason why physicists and chemists quickly embraced Schrödinger’s approach to quantum theory.

It is often overlooked, however, that the **first** formulation of quantum mechanics was not Schrödinger’s, but rather **Heisenberg’s**, and it emerged from a very different—and from today's point of view, much more abstract—starting point. Historically, Heisenberg’s approach was the culmination of efforts by many principally led by Niels Bohr and his school to explain atomic phenomena, particularly atomic spectra.

Bohr introduced the concept of **stationary states** and postulated without a good physical reason that systems in such states do not radiate. Radiation occurs only during **transitions** between these states, labeled by quantum numbers $m$ and $n$, and the frequency of the emitted or absorbed radiation is given by the **Bohr frequency condition**:

$$
\omega_{mn} = \frac{E_m - E_n}{\hbar}.
$$

Following Bohr’s proposal, experimentalists around the world began to study atomic spectra in great detail. The results of these efforts were tables of transition frequencies $\omega_{mn}$ and corresponding intensities, which we can denote by $P_{mn}$.

At the time, theoreticians attempted to explain these findings by combining classical mechanics with quantization rules (such as the Sommerfeld-Wilson conditions), which selected certain classical orbits as “allowed.” While this approach successfully reproduced energy levels of some simple systems, it ultimately failed to explain the **structure** of atomic spectra and in particular the intensities of various spectral lines. Here is where young Werner Heisenberg comes into play...

Heisenberg, a deep thinker with a philosophical inclination, arrived at a radical conclusion: on the atomic scale, a classical trajectory (for example, of an electron in a hydrogen atom) is **not observable**, since any attempt to measure it would disturb the system. Therefore, a theory of atomic phenomena should be built **only on observable quantities**.

And what were the observables? By the 1920s, the scientific community was already immersed in vast tables of transition frequencies $\omega_{mn}$ and their intensities $P_{mn}$. Inspired by this, Heisenberg proposed to **replace classical quantities** like the trajectory $x(t)$ by a **table of numbers**, each entry being associated with a pair of Bohr's stationary states and describing a transition between two stationary states:

$$
x(t) \longrightarrow x_{mn} e^{-i\omega_{mn} t}.
$$

Each entry $x_{mn}$ was assumed to oscillate with the frequency of the corresponding transition. Since absorption and stimulated emission rates were known to be symmetric, Heisenberg further postulated:

$$
x_{nm} = x_{mn}^* e^{i \omega_{mn} t},
$$

implying a particular symmetry between elements.

He reasoned that the same should hold for the momentum:

$$
p(t) \longrightarrow p_{mn} e^{-i\omega_{mn} t}.
$$

At this point, Heisenberg asked how to define derived quantities like $x^2$, $p^2$ that are needed to construct the energy (the Hamiltonian) of the system. The first naive element-wise squaring, such as

$$
(x^2)_{mn} = x_{mn}^2 e^{-2i\omega_{mn} t},
$$

fails in this respect, since it would result in transition frequencies that do not match observations. Instead, Heisenberg noted that the **Ritz combination principle**, known from very early days of the atomic spectroscopy, dictates that transition frequencies combine as:

$$
\omega_{mn} = \omega_{ml} + \omega_{ln}.
$$

This motivated the definition of the multiplication for $x$ as:

$$
(x^2)_{mn} e^{-i\omega_{mn} t} = \sum_l x_{ml} e^{-i\omega_{ml} t} \cdot x_{ln} e^{-i\omega_{ln} t},
$$

or simply:

$$
(x^2)_{mn} = \sum_l x_{ml} x_{ln}.
$$

We now immediately recognize this as the rule for **matrix multiplication** since we are trained to do so. However, in Heisenberg's times this was an exotic mathematics now well known to all physicists so Heisenberg did not immediately recognize the structure behind his multiplication rules. 

The same holds for momentum, where one starts from the classical identity:

$$
p = m\dot{x} \quad \Rightarrow \quad p_{mn} = m\dot{x}_{mn} = -im\omega_{mn} x_{mn}.
$$

From here, the Hamiltonian of the **harmonic oscillator** is introduced in its classical form:

$$
H = \frac{p^2}{2m} + \frac{1}{2} m \omega^2 x^2,
$$

and a corresponding “matrix” Hamiltonian is defined:

$$
H_{mn} e^{-i\omega_{mn} t},
$$

Since the Hamiltonian in the classical mechanics is a constant of motion, Heisenberg went on to postulate that his "table" extension should also be time-independent. However, this will be only the case if the $H$ is diagonal:

$$
H_{mn} = 
\begin{cases}
0 & m \neq n, \\
E_n & m = n,
\end{cases}
$$

Using the above one can can now construct the Hamiltonian for the harmonic oscillator, which will have the general form as:

$$
\begin{aligned}
H_{mn} &= \frac{1}{2m} \sum_l p_{ml} p_{ln} + \frac{1}{2} m \omega^2 \sum_l x_{ml} x_{ln} \\
&= \frac{1}{2} m \sum_l x_{ml} x_{ln} \left( \omega^2 - \omega_{ml} \omega_{ln} \right).
\end{aligned}
$$

To find a set of $x_{mn}$ such that this Hamiltonian is diagonal, Heisenberg was let to the following structure for $x$:

$$
\begin{aligned}
x_{n,n+1} &= \sqrt{\frac{(n+1)\hbar}{2m\omega}}, \\
x_{n+1,n} &= \sqrt{\frac{(n+1)\hbar}{2m\omega}},
\end{aligned}
$$

with all other elements vanishing. Inserting these into the expression for $H$, one readily obtains the **quantized energy levels** of the harmonic oscillator:

$$
H_{nn} = \left(n + \frac{1}{2}\right)\hbar \omega.
$$

From the same matrix $X$, one can now construct $P$ using:

$$
p_{n,n+1} = -i m \omega x_{n,n+1} = -i \sqrt{\frac{(n+1)\hbar m \omega}{2}}.
$$

After Heisenberg published these results in paper entitled 

*Zeitschrift für Physik*, 33, 879 (1925), his mentor Max Born recognized that Heisenberg’s “tables” were in fact **matrices**, and that the multiplication rule was matrix multiplication. From the explicit forms of $X$ and $P$, Born calculated their commutator and obtained:

$$
[X, P] = i\hbar,
$$

which became the canonical commutation relation of quantum mechanics.

### The Heisenberg Recipe for Solving Quantum Problems

In the original matrix formulation of quantum mechanics, solving a problem proceeds as follows:

1. **Construct matrices $X$ and $P$** such that

   $$
   [X, P] = i\hbar.
   $$

   This relation defines the fundamental structure of the quantum phase space and encodes the uncertainty principle at the operator level.

2. **Construct the Hamiltonian $H(X, P)$** and diagonalize it. The diagonal elements of $H$ yield the **energy levels** $E_n$ of the system. That is, find a unitary transformation $U$ such that:

   $$
   U^\dagger H U = \text{diag}(E_0, E_1, \dots),
   $$

   and the columns of $U$ are the eigenstates of the system.

This is the essence of **matrix mechanics**—a formulation that does not rely on wavefunctions, but rather on algebraic relations between observables and their transition properties.
