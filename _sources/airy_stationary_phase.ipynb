## ðŸ”¸ Stationary Phase Method applied to Airy function

We consider the integral representation of the solution of the SchrÃ¶dinger equation in a linear potential (assuming $m=1, \hbar =1, E=0$. In this case the wavefunction is given by a following Fourier integral:

$$
\psi(x) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{+\infty} e^{i\left(px + \frac{p^3}{6F}\right)} \, dp,
$$

In the following, we investigate its asymptotic behavior for large negative values of $ x $, i.e., $ x \to -\infty $. This type of integral is typical in semiclassical analysis and appears in the context of the semiclassical WKB method, Airy functions and wavepacket propagation.


### Phase Function and Stationary Points

We define the phase function

$$
g(p) = px + \frac{p^3}{6F},
$$

so that the integrand becomes $ e^{i g(p)} $. The method of stationary phase is based on the idea that the largest contribution to the integral comes from the points at which the phase is stationary since away from such points fast oscillations develop that integrate to zero in the asymptotic limit. Thus in order to evaluate the itnegral using the method of stationary phase we need to locate the stationary points of $ g(p) $, i.e., the points where $ g'(p) = 0 $:

$$
g'(p) = x + \frac{p^2}{2F} = 0.
$$

Solving this equation for $ x < 0 $, we find two real stationary points:

$$
p_\pm = \pm \sqrt{-2F x} = \pm \sqrt{2F |x|}.
$$

We now evaluate the phase function at the stationary points. For $ p_+ = \sqrt{2F |x|} $, we have

$$
\begin{aligned}
g(p_+) &= p_+ x + \frac{p_+^3}{6F} \\
&= -\sqrt{2F} |x|^{3/2} + \frac{(2F)^{3/2}}{6F} |x|^{3/2} \\
&= \left( -1 + \frac{\sqrt{2F}}{3F} \right) \sqrt{2F} |x|^{3/2} = \frac{4}{3} \sqrt{2F} |x|^{3/2}.
\end{aligned}
$$

By symmetry, we also have
$$
g(p_-) = -\frac{4}{3} \sqrt{2F} |x|^{3/2}.
$$


The second derivative at the stationary points is

$$
g''(p) = \frac{p}{F}, \quad \Rightarrow \quad g''(p_\pm) = \pm \sqrt{\frac{2}{F}} |x|^{1/2}.
$$

Hence, the magnitude of the second derivative is the same at both saddle points:

$$
|g''(p_\pm)| = \sqrt{\frac{2}{F}} |x|^{1/2}.
$$


### Application of the Stationary Phase Approximation

According to the stationary phase approximation for non-degenerate critical points, we have

$$
\int e^{i g(p)} \, dp \approx \sum_{\text{stationary } p_0} e^{i g(p_0)} \sqrt{\frac{2\pi}{|g''(p_0)|}} \, e^{i \sigma_{p_0} \frac{\pi}{4}},
$$

where $ \sigma_{p_0} = \operatorname{sign}(g''(p_0)) $.

In our case, this gives:

$$
\begin{aligned}
\psi(x) &\approx \frac{1}{\sqrt{2\pi}} \left[
\sqrt{\frac{2\pi}{|g''(p_+)|}} e^{i g(p_+) - i\pi/4} +
\sqrt{\frac{2\pi}{|g''(p_-)|}} e^{i g(p_-) + i\pi/4}
\right] \\
&= \sqrt{\frac{1}{|g''(p_+)|}} \left[
e^{i g(p_+) - i\pi/4} +
e^{-i g(p_+) + i\pi/4}
\right] \\
&= 2 \sqrt{\frac{1}{|g''(p_+)|}} \cos\left( g(p_+) - \frac{\pi}{4} \right).
\end{aligned}
$$


We finally substitute the expressions for \( g(p_+) \) and \( |g''(p_+)| \):
$$
g(p_+) = \frac{4}{3} \sqrt{2F} |x|^{3/2}, \qquad
|g''(p_+)| = \sqrt{\frac{2}{F}} |x|^{1/2}.
$$

Thus, the asymptotic form of the wavefunction becomes
$$
\psi(x) \approx 2 \left( \frac{F}{2} \right)^{1/4} \cdot \frac{1}{|x|^{1/4}} \cdot
\cos\left( \frac{4}{3} \sqrt{2F} |x|^{3/2} - \frac{\pi}{4} \right).
$$


$$
\boxed{
\psi(x) \approx 2 \left( \frac{F}{2} \right)^{1/4} \frac{1}{|x|^{1/4}}
\cos\left( \frac{4}{3} \sqrt{2F} |x|^{3/2} - \frac{\pi}{4} \right)
\quad \text{for } x \to -\infty
}
$$




## ðŸ”¹  Justification and Error Estimates for the Stationary Phase Approximation

In order to justify our approach a bit further, we consider a general oscillatory integral of the form
$$
I(x) = \int_{-\infty}^{+\infty} \, e^{i x g(p)} \, dp,
$$
where $ x \gg 1 $ is a large parameter. We assume for simplicity that $ g(p) $ is a smooth real-valued function whose stationary points (where $ g'(p) = 0 $) determine the dominant contributions to the integral.


When $x$  is large, the integrand oscillates rapidly except near points where the phase is stationary. Further away from these points, the contributions to the integral largely cancel out due to a highly oscillatory behaviour. The leading contribution to the integral thus comes from small neighborhoods around the stationary points, where the phase varies rather slowly.

Suppose $ p_0 $ is a non-degenerate stationary point, i.e.,

$$
g'(p_0) = 0, \quad g''(p_0) \neq 0.
$$

Expanding $ g(p) $ in a Taylor series around $ p_0 $:

$$
g(p) = g(p_0) + \frac{1}{2} g''(p_0)(p - p_0)^2 + \cdots.
$$

Using this expansion the integral becomes locally

$$
\int e^{i x \left(g(p_0) + \frac{1}{2} g''(p_0)(p - p_0)^2\right)} \, dp.
$$

This is a standart Gaussian integral, and its leading-order contribution is

$$
I(x) \approx  e^{i \lambda g(p_0)} \sqrt{\frac{2\pi}{x |g''(p_0)|}} e^{i \sigma \frac{\pi}{4}},
$$

where $ \sigma = \operatorname{sign}(g''(p_0)) $. If there are multiple isolated stationary points, their contributions add up.


The error term in the stationary phase approximation depends on how rapidly the phase varies away from the stationary point. In the case of a single non-degenerate stationary point,

$$
\int f(p) e^{i \lambda g(p)} \, dp = f(p_0) e^{i \lambda g(p_0)} \sqrt{\frac{2\pi}{\lambda |g''(p_0)|}} e^{i \sigma \frac{\pi}{4}} + \mathcal{O}(\lambda^{-3/2}).
$$

The higher-order terms involve higher derivatives $ g $. One can show that the integral has an asymptotic expansion of the form

$$
I(\lambda) \sim \sum_{k=0}^{n} \frac{A_k}{\lambda^{k + 1/2}} + \mathcal{O}(\lambda^{-n - 3/2}),
$$

with some coefficients $ A_k $.
