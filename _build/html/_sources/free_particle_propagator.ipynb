
## Free Particle Propagator

We now consider the **time evolution operator** in the position representation. As a starting point, we focus on the **free particle**, whose Hamiltonian is given by

$$
\hat{H} = \frac{\hat{p}^2}{2m}.
$$

Since the Hamiltonian is time-independent, the time evolution operator for a time interval $t$ is obtained by exponentiation:

$$
\hat{U}(t) = e^{-\frac{i}{\hbar} \hat{H} t}.
$$

The quantum state at time $t$ evolves from the initial state $|\psi_0\rangle$ via

$$
|\psi(t)\rangle = \hat{U}(t) |\psi_0\rangle.
$$

To express this in the position representation, we project onto a position eigenstate $|x\rangle$, yielding the wave function:

$$
\psi(x, t) := \langle x | \psi(t) \rangle = \langle x | \hat{U}(t) | \psi_0 \rangle.
$$

We now insert the identity operator in the form of a completeness relation over the position eigenstates:

$$
\psi(x, t) = \int_{-\infty}^{+\infty} dx' \, \langle x | \hat{U}(t) | x' \rangle \langle x' | \psi_0 \rangle.
$$

This motivates the definition of the **propagator kernel** (or Green's function):

$$
K(x, x'; t) := \langle x | \hat{U}(t) | x' \rangle,
$$

so that the time-evolved wave function can be written as

$$
\psi(x, t) = \int_{-\infty}^{+\infty} dx' \, K(x, x'; t) \, \psi(x', 0).
$$

---

### Evaluation of the Propagator $K(x, x'; t)$

To evaluate $K(x, x'; t)$ for the free particle, it is convenient to use the **momentum eigenbasis**, which diagonalizes the Hamiltonian. We insert a resolution of identity in terms of momentum eigenstates:

$$
\begin{aligned}
K(x, x'; t) &= \langle x | e^{-\frac{i}{\hbar} \frac{\hat{p}^2}{2m} t} | x' \rangle \\
&= \int_{-\infty}^{+\infty} \frac{dp}{2\pi\hbar} \, \langle x | e^{-\frac{i}{\hbar} \frac{\hat{p}^2}{2m} t} | p \rangle \langle p | x' \rangle.
\end{aligned}
$$

Using the fact that momentum eigenstates are eigenfunctions of $\hat{p}$ and that

$$
\langle x | p \rangle = \frac{1}{\sqrt{2\pi\hbar}} e^{\frac{i}{\hbar} p x}, \qquad \langle p | x' \rangle = \frac{1}{\sqrt{2\pi\hbar}} e^{-\frac{i}{\hbar} p x'},
$$

we obtain:

$$
\begin{aligned}
K(x, x'; t) &= \int_{-\infty}^{+\infty} \frac{dp}{2\pi\hbar} \, e^{-\frac{i}{\hbar} \frac{p^2}{2m} t} \cdot e^{\frac{i}{\hbar} p(x - x')} \\
&= \frac{1}{2\pi\hbar} \int_{-\infty}^{+\infty} dp \, \exp\left[\frac{i}{\hbar} \left( p(x - x') - \frac{p^2}{2m} t \right) \right].
\end{aligned}
$$

This is a **Gaussian integral** in $p$, and can be evaluated explicitly. The result is

$$
K(x, x'; t) = \sqrt{\frac{m}{2\pi i \hbar t}} \, \exp\left[ \frac{i m (x - x')^2}{2\hbar t} \right],
$$

which is the **free particle propagator** in one dimension. It expresses the amplitude for a particle initially localized at $x'$ to be found at position $x$ after time $t$, under free evolution.




## Path Integral Representation of the Propagator

We now derive the path integral form of the propagator for a quantum particle. The time evolution operator is

$$
\hat{U}(t) = e^{-\frac{i}{\hbar} \hat{H} t}, \quad \text{so that} \quad K(x, x'; t) = \langle x | \hat{U}(t) | x' \rangle.
$$

To evaluate this, we divide the time interval $[0, t]$ into $N$ small steps of length $\Delta t = t / N$. Using the composition property of the time-evolution operator:

$$
\hat{U}(t) = \hat{U}(t_N, t_{N-1}) \cdots \hat{U}(t_2, t_1) \hat{U}(t_1, t_0),
$$

we write the propagator as a product of infinitesimal propagators:

$$
K(x, x'; t) = \langle x | \hat{U}(t_N, t_{N-1}) \cdots \hat{U}(t_1, t_0) | x' \rangle.
$$

We now insert resolutions of the identity in the position basis between each time slice:

$$
\mathbb{I} = \int dx_i \, |x_i\rangle \langle x_i|, \quad \text{for } i = 1, \dots, N-1.
$$

This yields:

$$
K(x, x'; t) = \int dx_1 \cdots dx_{N-1} \prod_{j=1}^{N} \langle x_j | \hat{U}(\Delta t) | x_{j-1} \rangle,
$$

where $x_N = x$ and $x_0 = x'$.

For small $\Delta t$, and assuming a Hamiltonian of the form

$$
\hat{H} = \frac{\hat{p}^2}{2m} + V(\hat{x}),
$$

we approximate the short-time propagator using Trotter splitting:

$$
\hat{U}(\Delta t) \approx e^{-\frac{i}{\hbar} \hat{H} \Delta t} \approx e^{-\frac{i}{\hbar} \frac{\hat{p}^2}{2m} \Delta t} e^{-\frac{i}{\hbar} V(\hat{x}) \Delta t} + \mathcal{O}(\Delta t^2).
$$

The matrix elements of the kinetic term in the position basis are known:

$$
\langle x_j | e^{-\frac{i}{\hbar} \frac{\hat{p}^2}{2m} \Delta t} | x_{j-1} \rangle = \sqrt{\frac{m}{2\pi i \hbar \Delta t}} \exp\left[ \frac{i m}{2 \hbar \Delta t} (x_j - x_{j-1})^2 \right].
$$

Multiplying by the potential term (which is diagonal in position space):

$$
\langle x_j | \hat{U}(\Delta t) | x_{j-1} \rangle \approx \sqrt{\frac{m}{2\pi i \hbar \Delta t}} \exp\left[ \frac{i}{\hbar} \left( \frac{m}{2} \left( \frac{x_j - x_{j-1}}{\Delta t} \right)^2 - V(x_{j-1}) \right) \Delta t \right].
$$

Putting it all together:

$$
K(x, x'; t) = \lim_{N \to \infty} \left( \frac{m}{2\pi i \hbar \Delta t} \right)^{N/2} \int dx_1 \cdots dx_{N-1} \exp\left[ \frac{i}{\hbar} \sum_{j=1}^{N} \left( \frac{m}{2} \left( \frac{x_j - x_{j-1}}{\Delta t} \right)^2 - V(x_{j-1}) \right) \Delta t \right].
$$

In the continuum limit, this becomes a functional integral over all possible paths $x(t)$ from $x(0) = x'$ to $x(t) = x$:

$$
K(x, x'; t) = \int \mathcal{D}[x(t)] \, e^{\frac{i}{\hbar} \int_0^t dt \, L(x, \dot{x})},
$$

where $L(x, \dot{x}) = \frac{1}{2} m \dot{x}^2 - V(x)$ is the classical Lagrangian. This is the **Feynman path integral formulation** of quantum mechanics.
