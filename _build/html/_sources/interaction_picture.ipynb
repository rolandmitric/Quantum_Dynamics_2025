# Interaction Picture of Quantum Mechanics

We have seen that the transition from the Schrödinger picture to the Heisenberg picture completely removes the time-dependence from the state vector and shifts it to the observables (operators). If $\left|\psi_s(t)\right\rangle$ is the state vector in the Schrödinger picture, then

$$
\begin{aligned}
& \left|\psi_s(t)\right\rangle = U(t) \left|\psi_H(t)\right\rangle \\
& \left|\psi_H(t)\right\rangle = U^{\dagger}(t) \left|\psi_s(t)\right\rangle
\end{aligned}
$$

The expectation values in the Schrödinger picture

$$
\langle A \rangle = \left\langle \psi_s \middle| A_s \middle| \psi_s \right\rangle
$$

can be re-expressed as the expectation values in the Heisenberg picture via

$$
\langle A \rangle = \left\langle \psi_s \middle| A_s \middle| \psi_s \right\rangle = \left\langle \psi_H \middle| U^{\dagger} A_s U \middle| \psi_H \right\rangle
$$

which motivates the definition of the Heisenberg operator

$$
A_H = U^{\dagger} A U
$$

---

We now introduce a third picture which is particularly useful when the Hamiltonian of the system can be split into two parts, one of which is time-independent:

$$
H = H_0 + V(t)
$$

Typically in chemical applications, $H_0$ would be the molecular or atomic Hamiltonian and $V(t)$ represents the interaction with some external time-dependent field (e.g. the electric field of a light wave). In such cases, it would be desirable to transform away the "trivial" part of the time-dependence induced by $H_0$ by moving to a new “rotating” frame of reference. This is the idea of the interaction picture of quantum mechanics, first introduced by Dirac.

From our treatment of the rotating-wave approximation, we recall that we can transform the time evolution operator $U(t)$ to a new "rotating" system using an arbitrary unitary operator $R(t)$ such that

$$
U(t) = \left(R R^{\dagger}\right) U(t) = R(t) \tilde{U}(t)
$$

This transformation generally modifies the TDSE into an equation of motion for $\tilde{U}(t)$ as follows:

$$
U(t) = R(t) \tilde{U}(t) \quad \Rightarrow \quad \frac{dU}{dt} = \frac{dR}{dt} \tilde{U} + R \frac{d\tilde{U}}{dt}
$$

Substituting into the TDSE:

$$
i\hbar \frac{dU}{dt} = H U \quad \Rightarrow \quad i\hbar \left( \frac{dR}{dt} \tilde{U} + R \frac{d\tilde{U}}{dt} \right) = H R \tilde{U}
$$

From this it follows that:

$$
\begin{aligned}
\frac{d\tilde{U}}{dt} &= R^{\dagger} H R \tilde{U} - i\hbar R^{\dagger} \frac{dR}{dt} \tilde{U} \\
&= H_{\text{eff}}(t) \tilde{U}
\end{aligned}
$$

with the effective Hamiltonian:

$$
H_{\text{eff}}(t) = R^{\dagger} H R - i\hbar R^{\dagger} \frac{dR}{dt}
$$

We now ask: What is the best choice of $R(t)$ for a system with a Hamiltonian $H = H_0 + V(t)$ such that the effective Hamiltonian simplifies optimally?

Let us examine:

$$
\begin{aligned}
H_{\text{eff}}(t) &= R^{\dagger} \left( H_0 + V \right) R - i\hbar R^{\dagger} \frac{dR}{dt} \\
&= R^{\dagger} H_0 R + R^{\dagger} V R - i\hbar R^{\dagger} \frac{dR}{dt} \\
&= R^{\dagger} \left[ H_0 R - i\hbar \frac{dR}{dt} \right] + R^{\dagger} V R
\end{aligned}
$$

We now see that if we choose $R(t)$ to satisfy the TDSE for the free Hamiltonian:

$$
i\hbar \frac{dR}{dt} = H_0 R
$$

then the first term vanishes, and the effective Hamiltonian becomes:

$$
H_{\text{eff}}(t) = R^{\dagger} V R
$$

Thus, choosing $R$ to be the propagator of the unperturbed system effectively transforms away the time dependence induced by $H_0$. Since $H_0$ is time-independent, we can write:

$$
R(t) = e^{-i H_0 t / \hbar}
$$

and define the **interaction picture propagator** $U_I(t)$ as:

$$
U_I(t) := \tilde{U}(t) = R^{\dagger}(t) U(t)
$$

or explicitly:

$$
U_I(t) = e^{i H_0 t / \hbar} U(t)
$$

The equation of motion for $U_I(t)$ becomes:

$$
i\hbar \frac{dU_I}{dt} = R^{\dagger} V R \cdot U_I(t)
$$

We define:

$$
V_I(t) := R^{\dagger} V R = e^{i H_0 t / \hbar} V(t) e^{-i H_0 t / \hbar}
$$

This is called the interaction picture representation of the operator $V(t)$. Thus, the equation for the interaction picture propagator reads:

$$
i\hbar \frac{dU_I}{dt} = V_I(t) U_I(t)
$$

The intuition is that the transition to the interaction picture removes the rapid oscillations induced by $H_0$ and leaves only the more slowly varying dynamics caused by $V(t)$. This separation is meaningful only when the operator norm of $V(t)$ is small compared to $H_0$—a condition that justifies the use of time-dependent perturbation theory.

---

# Time-Dependent Perturbation Theory

Let us now assume that after transforming to the interaction picture, our problem is reduced to solving:

$$
i\hbar \frac{dU_I}{dt} = V_I(t) U_I(t)
$$

This is still a complicated problem, and an exact solution is generally not possible. The idea of perturbation theory is to convert one hard problem into infinitely many simpler ones. We introduce a small formal parameter $\varepsilon$:

$$
i\hbar \frac{dU_I}{dt} = \varepsilon V_I(t) U_I(t)
$$

This makes $U_I(t)$ now $\varepsilon$-dependent:

$$
U_I(t) \to U_I(t, \varepsilon)
$$

We assume $U_I(t, \varepsilon)$ is analytic in $\varepsilon$, and expand it into a power series:

$$
U_I(t, \varepsilon) = \sum_{n=0}^{\infty} \varepsilon^n U_I^{(n)}(t)
$$

Inserting this into the TDSE:

$$
\frac{dU_I}{dt} = \sum_{n=0}^\infty \varepsilon^n \frac{dU_I^{(n)}}{dt}
$$

We obtain:

$$
i\hbar \sum_{n=0}^\infty \varepsilon^n \frac{dU_I^{(n)}}{dt} = \varepsilon V_I(t) \sum_{n=0}^\infty \varepsilon^n U_I^{(n)}(t)
$$

On the right-hand side, we shift the summation index:

$$
\sum_{n=0}^\infty \varepsilon^n \left( i\hbar \frac{dU_I^{(n)}}{dt} \right) = \sum_{n=1}^\infty \varepsilon^n V_I(t) U_I^{(n-1)}(t)
$$

Matching powers of $\varepsilon$, we find:

$$
\begin{aligned}
& i\hbar \frac{dU_I^{(0)}}{dt} = 0 \\
& i\hbar \frac{dU_I^{(n)}}{dt} = V_I(t) U_I^{(n-1)}(t), \quad n = 1, 2, \ldots
\end{aligned}
$$

The zeroth-order equation gives:

$$
U_I^{(0)}(t) = \mathbb{1}
$$

Then:

$$
i\hbar \frac{dU_I^{(1)}}{dt} = V_I(t)
$$

which integrates to:

$$
U_I^{(1)}(t) = \frac{1}{i\hbar} \int_0^t d\tau_1 \, V_I(\tau_1)
$$

Proceeding recursively, we find:

$$
U_I^{(2)}(t) = \frac{1}{(i\hbar)^2} \int_0^t d\tau_2 \int_0^{\tau_2} d\tau_1 \, V_I(\tau_2) V_I(\tau_1)
$$

and in general:

$$
U_I^{(n)}(t) = \frac{1}{(i\hbar)^n} \int_0^t d\tau_n \int_0^{\tau_n} d\tau_{n-1} \cdots \int_0^{\tau_2} d\tau_1 \, V_I(\tau_n) \cdots V_I(\tau_1)
$$

Thus, the full Dyson series becomes:

$$
U_I(t) = \mathbb{1} + \sum_{n=1}^{\infty} \left( \frac{\varepsilon}{i\hbar} \right)^n \int_{0 < \tau_1 < \cdots < \tau_n < t} V_I(\tau_n) \cdots V_I(\tau_1) \, d\tau_1 \cdots d\tau_n
$$



## Example: Constant Perturbation — First-Order Time-Dependent Perturbation Theory

We consider a quantum system described by a time-independent unperturbed Hamiltonian $H_0$, and a constant (i.e., time-independent) perturbation $V$. The full Hamiltonian is:

$$
H = H_0 + \varepsilon V
$$

Let $ \{ |n\rangle \} $ be the orthonormal eigenbasis of $ H_0 $, such that:

$$
H_0 |n\rangle = E_n |n\rangle
$$

Suppose the system is initially in the eigenstate $|i\rangle$ at time $t = 0$. We want to compute the probability of finding the system in a different eigenstate $|f\rangle$ at a later time $t$, to first order in $ \varepsilon $).

---

### Step 1: Switch to the Interaction Picture

The interaction picture time-evolution operator satisfies:

$$
i\hbar \frac{dU_I(t)}{dt} = \varepsilon V_I(t) U_I(t)
$$

Since $V$ is constant and $H_0$ is time-independent, the interaction-picture perturbation becomes:

$$
V_I(t) = e^{i H_0 t/\hbar} V e^{-i H_0 t/\hbar}
$$

This is generally **not constant** in the interaction picture, even if $V$ is constant in the Schrödinger picture.

---

### Step 2: First-Order Expansion of the Propagator

We approximate the interaction-picture evolution operator to first order:

$$
U_I(t) \approx \mathbb{1} - \frac{i \varepsilon}{\hbar} \int_0^t d\tau \, V_I(\tau)
$$

Thus, the full time-evolution operator in the Schrödinger picture becomes:

$$
U(t) \approx e^{-i H_0 t / \hbar} \left( \mathbb{1} - \frac{i \varepsilon}{\hbar} \int_0^t d\tau \, V_I(\tau) \right)
$$

---

### Step 3: Transition Amplitude

We compute the transition amplitude from $|i\rangle$ to $|f\rangle$:

$$
\langle f | U(t) | i \rangle \approx \langle f | e^{-i H_0 t/\hbar} \left( \mathbb{1} - \frac{i \varepsilon}{\hbar} \int_0^t d\tau \, V_I(\tau) \right) |i\rangle
$$

Since $ |i\rangle$ and $|f\rangle $ are eigenstates of $H_0$, we have:

$$
e^{-i H_0 t/\hbar} |i\rangle = e^{-i E_i t/\hbar} |i\rangle \\
\langle f | e^{-i H_0 t/\hbar} = e^{-i E_f t/\hbar} \langle f |
$$

Using this, the matrix element becomes:

$$
\langle f | U(t) | i \rangle \approx e^{-i E_f t/\hbar} \left[ \delta_{fi} - \frac{i \varepsilon}{\hbar} \int_0^t d\tau \, \langle f | e^{i H_0 \tau / \hbar} V e^{-i H_0 \tau / \hbar} | i \rangle \right]
$$

The matrix element inside the integral becomes:

$$
\langle f | e^{i H_0 \tau / \hbar} V e^{-i H_0 \tau / \hbar} | i \rangle = e^{i (E_f - E_i) \tau / \hbar} \langle f | V | i \rangle
$$

Thus:

$$
\langle f | U(t) | i \rangle \approx e^{-i E_f t/\hbar} \left[ \delta_{fi} - \frac{i \varepsilon}{\hbar} \langle f | V | i \rangle \int_0^t d\tau \, e^{i (E_f - E_i) \tau / \hbar} \right]
$$

---

### Step 4: Evaluate the Time Integral

Define the energy difference:

$$
\omega_{fi} := \frac{E_f - E_i}{\hbar}
$$

Then:

$$
\int_0^t d\tau \, e^{i \omega_{fi} \tau} = \frac{e^{i \omega_{fi} t} - 1}{i \omega_{fi}} = e^{i \omega_{fi} t/2} \cdot \frac{2 \sin(\omega_{fi} t / 2)}{\omega_{fi}}
$$

So:

$$
\langle f | U(t) | i \rangle \approx e^{-i E_f t/\hbar} \left[ \delta_{fi} - \varepsilon \cdot \langle f | V | i \rangle \cdot \frac{e^{i \omega_{fi} t/2} \cdot 2 \sin(\omega_{fi} t / 2)}{\hbar \omega_{fi}} \right]
$$

---

### Step 5: Transition Probability

The transition probability is:

$$
P_{i \rightarrow f}(t) = |\langle f | U(t) | i \rangle|^2
$$

For $ f \neq i $, and to first order in $ \varepsilon$, this becomes:

$$
P_{i \rightarrow f}(t) \approx \left| \varepsilon \cdot \langle f | V | i \rangle \cdot \frac{2 \sin(\omega_{fi} t / 2)}{\hbar \omega_{fi}} \right|^2
$$

This gives the well-known result:

$$
P_{i \rightarrow f}(t) = \frac{4 \varepsilon^2 |\langle f | V | i \rangle|^2}{\hbar^2 \omega_{fi}^2} \cdot \sin^2\left(\frac{\omega_{fi} t}{2}\right)
$$

---

### Interpretation

- The transition probability oscillates with time.
- For long times, the function $ \frac{\sin^2(\omega t/2)}{(\omega/2)^2} $ becomes sharply peaked around $\omega = 0 $, i.e., $ E_f = E_i $.
- In the limit $ t \to \infty $, this becomes proportional to a delta function, leading to **Fermi’s Golden Rule**:

$$
\Gamma_{i \rightarrow f} = \frac{2\pi}{\hbar} \varepsilon^2 |\langle f | V | i \rangle|^2 \delta(E_f - E_i)
$$


## Time-Ordered Form of the Perturbative Expansion

### 1. Time Evolution in the Interaction Picture

Starting from the Dyson series, we now rewrite it by defining the so called the time-ordering operator. This is not that much useful for practical calculations that are of interest for us, but it yields a compact mathematical expression.

$$
\begin{aligned}
U_I(t, 0)
&= I - \frac{i}{\hbar} \int_0^t dt_1 \, V_I(t_1) 
+ \left( -\frac{i}{\hbar} \right)^2 \int_0^t dt_2 \int_0^{t_2} dt_1 \, V_I(t_2) V_I(t_1) \\
&\quad + \left( -\frac{i}{\hbar} \right)^3 \int_0^t dt_3 \int_0^{t_3} dt_2 \int_0^{t_2} dt_1 \, V_I(t_3) V_I(t_2) V_I(t_1) + \cdots
\end{aligned}
$$

This expression, written in terms of nested integrals over ordered time variables, becomes:

$$
U_I(t, 0) = \sum_{n=0}^\infty \left( -\frac{i}{\hbar} \right)^n \int_{0 \le t_1 \le \cdots \le t_n \le t} dt_n \cdots dt_1 \, V_I(t_n) \cdots V_I(t_1)
$$

---

### 2. Replacement by Integration over the Cube

To rewrite the Dyson expansion in a compact form, we define the **time-ordering operator** $\mathcal{T}$, which acts on operator products such that the time arguments are rearranged in decreasing order (i.e., later times to the left):

$$
\mathcal{T}[V_I(t_1) \cdots V_I(t_n)] = V_I(t_{\pi(1)}) \cdots V_I(t_{\pi(n)}) \quad \text{with } t_{\pi(1)} \ge \cdots \ge t_{\pi(n)}
$$

Then the identity

$$
\int_{0 \le t_1 \le \cdots \le t_n \le t} V_I(t_n) \cdots V_I(t_1) \, dt_n \cdots dt_1 = \frac{1}{n!} \int_{[0, t]^n} \mathcal{T}[V_I(t_1) \cdots V_I(t_n)] \, dt_n \cdots dt_1
$$

allows us to write the Dyson series in compact form:

$$
U_I(t, 0) = \mathcal{T} \exp\left( -\frac{i}{\hbar} \int_0^t V_I(t') \, dt' \right)
$$

---

## Justification of Time-Ordering Replacement

Let $f(t_1, \dots, t_n)$ be an integrable function. Define the **simplex**:

$$
\Delta = \{ (t_1, \dots, t_n) \in [0,t]^n \mid t_1 \le \cdots \le t_n \}
$$

Then:

$$
\int_\Delta f(t_1, \dots, t_n) \, dt_1 \cdots dt_n = \frac{1}{n!} \int_{[0,t]^n} \mathcal{T}[f(t_1, \dots, t_n)] \, dt_1 \cdots dt_n
$$

### Proof Sketch:

1. The cube $[0,t]^n$ can be partitioned into $n!$ disjoint regions $D_\pi$, one for each permutation $\pi \in S_n$, defined by:

$$
D_\pi := \{ (t_1, \dots, t_n) \in [0,t]^n \mid t_{\pi(1)} \le \cdots \le t_{\pi(n)} \}
$$

2. Each point in the cube belongs to exactly one such region, and on each region the function $f$ can be re-ordered to fit the standard simplex by applying $\mathcal{T}$.

3. Therefore:

$$
\int_{[0,t]^n} \mathcal{T}[f(t_1, \dots, t_n)] \, dt_1 \cdots dt_n 
= \sum_{\pi \in S_n} \int_{D_\pi} f(t_{\pi(1)}, \dots, t_{\pi(n)}) \, dt_1 \cdots dt_n 
= n! \int_\Delta f(t_1, \dots, t_n) \, dt_1 \cdots dt_n
$$

Dividing by $n!$ gives the result.

---

### Examples

These identities can be illustrated by working out low-order cases:

**For $n = 2$:**

We are integrating over the triangular region where $0 \le t_1 \le t_2 \le t$, and the operator product $V(t_2)V(t_1)$ is already time-ordered. The corresponding identity is:

$$
\int_0^t dt_2 \int_0^{t_2} dt_1 \, V(t_2)V(t_1)
= \frac{1}{2} \int_0^t dt_1 \int_0^t dt_2 \, \mathcal{T}[V(t_1)V(t_2)]
$$

The prefactor $\frac{1}{2}$ corrects for the fact that the triangular region is half the square $[0,t]^2$, and $\mathcal{T}$ ensures that the product is always time-ordered regardless of whether $t_1 > t_2$ or not.

**For $n = 3$:**

We integrate over the simplex $0 \le t_1 \le t_2 \le t_3 \le t$. The product $V(t_3)V(t_2)V(t_1)$ is time-ordered. The identity becomes:

$$
\int_0^t dt_3 \int_0^{t_3} dt_2 \int_0^{t_2} dt_1 \, V(t_3)V(t_2)V(t_1)
= \frac{1}{6} \int_0^t dt_1 dt_2 dt_3 \, \mathcal{T}[V(t_1)V(t_2)V(t_3)]
$$

Here, $\frac{1}{6} = \frac{1}{3!}$ accounts for the fact that the simplex occupies $1/6$ of the cube $[0,t]^3$. Again, the time-ordering operator guarantees the correct operator sequence.



```python
import matplotlib.pyplot as plt
import numpy as np

def plot_dyson_region_n2():
    t = 1
    x = np.linspace(0, t, 100)

    fig, ax = plt.subplots()
    ax.fill_between(x, 0, x, color="lightblue", alpha=0.6, label="t1 ≥ t2")
    ax.plot(x, x, color="blue", linestyle="--")
    ax.set_xlabel("t1")
    ax.set_ylabel("t2")
    ax.set_title("Integration Region for n = 2")
    ax.set_xlim(0, t)
    ax.set_ylim(0, t)
    ax.legend()
    ax.set_aspect("equal")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_dyson_region_n2()
```

```python
import numpy as np
import plotly.graph_objects as go

def plot_dyson_region_n3_interactive():
    # Create 3D grid
    t_vals = np.linspace(0, 1, 50)
    T1, T2 = np.meshgrid(t_vals, t_vals)
    T3 = np.minimum(T1, T2)  # Enforce t1 ≥ t2 ≥ t3

    # Create the surface plot
    fig = go.Figure(data=[
        go.Surface(
            x=T1,
            y=T2,
            z=T3,
            colorscale='jet',
            opacity=0.8,
            showscale=False
        )
    ])

    fig.update_layout(
        title="Integration Region for n = 3: t₁ ≥ t₂ ≥ t₃",
        scene=dict(
            xaxis_title='t₁',
            yaxis_title='t₂',
            zaxis_title='t₃',
            xaxis=dict(range=[0, 1]),
            yaxis=dict(range=[0, 1]),
            zaxis=dict(range=[0, 1]),
        ),
        margin=dict(l=0, r=0, b=0, t=40)
    )

    fig.show()

plot_dyson_region_n3_interactive()
```
