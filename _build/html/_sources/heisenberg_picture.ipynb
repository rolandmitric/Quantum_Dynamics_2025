

## üîπ Observables in Quantum Mechanics

As we have learned physical systems in quantum mechanics are described by state vectors $|\psi(t)\rangle$ in a complex Hilbert space $\mathcal{H}$, and the dynamics are governed by unitary time-evolution evolution. To extract physical information from these states‚Äîsuch as position, momentum, energy‚Äîwe need a framework for representing *measurable quantities* that are in quantum mechanics usually called **observables**.

In the following section we postulate how observables are mathematically represented and how the outcomes of measurements can be extracted given a quantum state $|\psi\rangle$.

### Axiom (Observable):

 An **observable** is represented by a **Hermitian operator** $A$ on the Hilbert space $\mathcal{H}$:

$$
 A = A^\dagger,
$$

where $A^\dagger$ is the adjoint of $A$.

The motivation for this postulate lies in the mathematical properties of Hermitian operators: they always have real eigenvalues (ensuring real-valued measurement results) and a complete set of orthogonal eigenvectors (allowing decomposition of any state). These features are mathematically guaranteed by the **spectral theorem**, which we now state and prove in the finite-dimensional case.


###  Theorem (Spectral Theorem):

Let $V$ be a finite-dimensional complex Hilbert space, and let $A: V \to V$ be a Hermitian operator. Then:

1. All eigenvalues of $A$ are real.
2. There exists an **orthonormal basis** of $V$ consisting of eigenvectors of $A$.
3. In this basis, $A$ is represented by a diagonal matrix with real entries.

---

### üî∏ Proof:

We first prove, that if an eigevector exists the corresponding eigenvalue is real. Let $\lambda \in \mathbb{C}$ be an eigenvalue of $A$, and let $|v\rangle \neq 0$ be the corresponding eigenvector:

$$
A|v\rangle = \lambda |v\rangle.
$$

Taking the inner product with \$|v\rangle\$ gives:

$$
\langle v| A |v\rangle = \lambda \langle v|v\rangle.
$$

Using Hermiticity and the antilinearity of the inner product in the first argument we get:

$$
\langle v| A |v\rangle = \langle A v|v\rangle =  \lambda^* \langle v|v\rangle.
$$

Since $\langle v|v\rangle > 0$, it follows that $\lambda = \lambda^\*$, i.e., the eigenvalue is real.


Let us now show that at least one eigenvalue exists. Select any nonzero vector $|v\rangle \in V$. Consider the finite set of vectors:

$$
|v\rangle, \quad A|v\rangle, \quad A^2|v\rangle, \quad \dots, \quad A^n|v\rangle,
$$

where $n = \dim V$. This set contains $n+1$ vectors in an $n$-dimensional space, so they must be linearly dependent. That is, there exist complex coefficients $a\_0, a\_1, \dots, a\_n \in \mathbb{C}$, not all zero, such that:

$$
(a_n A^n + a_{n-1} A^{n-1} + \cdots + a_1 A + a_0 I) |v\rangle = 0.
$$

Define the complex polynomial:

$$
p(z) = a_n z^n + a_{n-1} z^{n-1} + \cdots + a_1 z + a_0,
$$

so that:

$$
p(A) |v\rangle = 0.
$$

This means that \$|v\rangle\$ lies in the **kernel** of the operator $p(A)$. 

By the **Fundamental Theorem of Algebra**, the polynomial \$p(z)\$ factorizes over \$\mathbb{C}\$:

$$
p(z) = c (z - \lambda_1)(z - \lambda_2)\cdots(z - \lambda_m),
$$

where $c \neq 0$, $\lambda_k \in \mathbb{C}$, and $m \leq n$. Applying this factorization to \$A\$ gives:

$$
p(A) = c (A - \lambda_1 I)(A - \lambda_2 I)\cdots(A - \lambda_m I),
$$

so the equation $p(A) |v\rangle = 0$ becomes:

$$
(A - \lambda_1 I)(A - \lambda_2 I)\cdots(A - \lambda_m I) |v\rangle = 0.
$$

Now suppose that **none** of the factors $(A - \lambda_k I)$ annihilate $|v\rangle$, i.e., that $(A - \lambda_k I) |w\rangle \neq 0$ for all $k$. Then, we could recursively define:

$$
|v_1\rangle := (A - \lambda_m I) |v\rangle, \quad |v_2\rangle := (A - \lambda_{m-1} I) |v_1\rangle, \quad \dots
$$

and finally obtain a contradiction, because applying $p(A)$ to $|v\rangle$ would produce a nonzero vector. Hence, **at least one** of the factors $(A - \lambda_j I)$ must satisfy:

$$
(A - \lambda_j I)|w\rangle = 0
$$

for some nonzero $|w\rangle \in V$. That is, $|w\rangle$ is an eigenvector of $A$, with eigenvalue $\lambda_j \in \mathbb{C}$.

Since we already showed that all eigenvalues of Hermitian operators are real, we conclude that $\lambda_j \in \mathbb{R}$.



To construct an orthonormal eigenbasis,we proceed by induction on \$\dim V\$:

* For \$n = 1\$, any nonzero vector is an eigenvector.
* Suppose the theorem holds for \$\dim V = n - 1\$. Let \$|v\_1\rangle\$ be an eigenvector with eigenvalue \$\lambda\_1\$, and define:

  $$
  W := \{ |w\rangle \in V \mid \langle w | v_1 \rangle = 0 \}.
  $$

  One can show \$W\$ is invariant under \$A\$, and \$A|\_W\$ is Hermitian. By the induction hypothesis, \$A|\_W\$ has an orthonormal eigenbasis. Together with \$|v\_1\rangle\$, we obtain a full orthonormal eigenbasis of \$V\$.

---

### üî∏ Spectral Decomposition

Let \${ |v\_i\rangle }\_{i=1}^n\$ be an orthonormal eigenbasis of \$A\$, with corresponding eigenvalues \$\lambda\_i \in \mathbb{R}\$. Then:

$$
A = \sum_{i=1}^n \lambda_i |v_i\rangle \langle v_i|,
$$

where each \$P\_i := |v\_i\rangle \langle v\_i|\$ is a **projection operator**.

In general, when eigenvalues are degenerate and eigenspaces have dimension > 1, we write:

$$
A = \sum_k \lambda_k P_k,
$$

where the \$P\_k\$ are **orthogonal projection operators** satisfying:

* \$P\_k^2 = P\_k\$ (idempotent),
* \$P\_k^\dagger = P\_k\$ (Hermitian),
* \$P\_j P\_k = \delta\_{jk} P\_k\$ (orthogonality),
* \$\sum\_k P\_k = I\$ (completeness).

The set containing all operators \${P\_k}\$ is called a **complete set of orthogonal projectors**.

---

## üîπ Quantum Measurement

We now formalize the notion of measurement in quantum mechanics.

###  Axiom (Measurement Postulate):

Let \${P\_k}\$ be a complete set of orthogonal projectors associated with the Hermitian observable \$A = \sum\_k \lambda\_k P\_k\$.

Then:

* When measuring \$A\$ in the state \$|\psi\rangle\$, the probability of obtaining outcome \$\lambda\_k\$ is:

$$
p(k) = \langle \psi | P_k | \psi \rangle = \| P_k |\psi\rangle \|^2.
$$

* Immediately after the measurement, the system is projected into the subspace \$\mathcal{H}\_k = \text{Im}(P\_k)\$:

$$
|\psi\rangle_{\text{after}} = \frac{P_k |\psi\rangle}{\| P_k |\psi\rangle \|}.
$$

---

## üîπ Expectation Value of an Observable

The **expectation value** (or average outcome) of an observable \$A\$ in the state \$|\psi\rangle\$ is:

$$
\langle A \rangle = \sum_k \lambda_k \, p(k) = \sum_k \lambda_k \langle \psi | P_k | \psi \rangle.
$$

Using the spectral decomposition of \$A\$, we also have:

$$
\langle A \rangle = \langle \psi | A | \psi \rangle.
$$




## üîπ Heisenberg Picture

We have seen that the quantum evolution of a system is described by a mapping from the real time axis into the Hilbert space $\mathcal{H}\$, such that each instant of time is associated with a state vector \$|\psi(t)\rangle \in \mathcal{H}\$. It is important to emphasize that the state vectors themselves do not *contain* time; instead, time appears as a parameter that labels a continuous family of states visited during the evolution:

$$
t \mapsto |\psi(t)\rangle \in \mathcal{H}
$$

This mapping defines a trajectory through Hilbert space, and the unitary time-evolution postulate guarantees that this trajectory is uniquely determined by the initial state \$|\psi(t\_0)\rangle\$. Each state along the trajectory can be generated from the initial state by applying a unitary time-evolution operator:

$$
|\psi(t)\rangle = U(t, t_0)\,|\psi(t_0)\rangle
$$

Hence, the family of states

$$
\left\{ |\psi(t)\rangle \,\middle|\, |\psi(t_0)\rangle \text{ given} \right\}
$$

is fully determined by the unitary dynamics.

Now we ask: **How do expectation values of quantum observables depend on time?**

Given a time-evolved state \$|\psi(t)\rangle\$, the expectation value of an observable \$A\$ at time \$t\$ is given by:

$$
\langle A(t)\rangle = \langle \psi(t) | A | \psi(t) \rangle
$$

Assuming the initial state \$|\psi(t\_0)\rangle\$ is normalized, unitarity ensures that \$|\psi(t)\rangle\$ remains normalized for all times.

Using the expression \$|\psi(t)\rangle = U(t, t\_0) |\psi(t\_0)\rangle\$, we can write the dual vector as:

$$
\langle \psi(t)| = \langle \psi(t_0) | U^\dagger(t, t_0)
$$

Thus, the time-dependent expectation value becomes:

$$
\langle A(t)\rangle = \langle \psi(t_0) | U^\dagger(t, t_0)\, A\, U(t, t_0) | \psi(t_0) \rangle
$$

So far, we have described quantum dynamics in the **Schr√∂dinger picture**, where the *state vectors* evolve in time while the *observables* (operators like \$A\$) remain time-independent elements of the operator algebra \$\mathscr{L}(\mathcal{H})\$.

The form of the expectation value above motivates a change in perspective. Instead of viewing the states as evolving, we can **transfer the time dependence to the operators** themselves.




Since the state of the system at any time \$t\$ is uniquely determined by the initial state \$|\psi(t\_0)\rangle\$, this initial state provides a complete description of the system's dynamics. The expression for the expectation value \$\langle A(t) \rangle\$ allows us to "shift" the time dependence from the state vectors to the operators themselves:

$$
\begin{aligned}
\langle A(t) \rangle 
&= \langle \psi_S(t) | A | \psi_S(t) \rangle \\
&= \langle \psi_S(t_0) | U^\dagger(t, t_0)\, A\, U(t, t_0) | \psi_S(t_0) \rangle \\
&= \langle \psi_H(t_0) | A_H(t) | \psi_H(t_0) \rangle
\end{aligned}
$$

Here, we have introduced the **Heisenberg-picture operator** and the corresponding state:

$$
\begin{aligned}
A_H(t) &= U^\dagger(t, t_0)\, A\, U(t, t_0) \\
|\psi_H(t_0)\rangle &= |\psi_S(t_0)\rangle
\end{aligned}
$$

Thus, in the **Heisenberg picture**, the *state vector* \$|\psi\_H(t\_0)\rangle\$ remains constant in time, while the *operators* evolve. Although mathematically equivalent to the Schr√∂dinger picture, the Heisenberg formulation is often conceptually closer to classical mechanics. In classical dynamics, we associate time-dependent functions \$x(t)\$ and \$p(t)\$ with a particle‚Äôs position and momentum. Similarly, in the Heisenberg picture, these observables become time-dependent operators \$x\_H(t)\$ and \$p\_H(t)\$.



Since the state of the system at any time \$t\$ is uniquely determined by \$\left|\psi\left(t\_0\right)\right\rangle\$, the vector \$\left|\psi\left(t\_0\right)\right\rangle\$ provides a full description of the system. The expression for \$\langle A\rangle\$ allows us to "shift" the time dependence from the state vectors to the operators in the following way:

$$
\begin{aligned}
\langle A\rangle & =\left\langle\psi_S(t)\right| A\left|\psi_S(t)\right\rangle=\left\langle\psi_S\left(t_0\right)\right| U^{+} A U\left|\psi_S\left(t_0\right)\right\rangle \\
& =\left\langle\psi_H\left(t_0\right)\right| A_H(t)\left|\psi_H\left(t_0\right)\right\rangle
\end{aligned}
$$

where we have defined the Heisenberg-picture operator

$$
\begin{gathered}
A_H(t)=U^{+}\left(t, t_0\right) A U\left(t, t_0\right) \\
\text{and} \\
\left|\psi_H\left(t_0\right)\right\rangle=\left|\psi_S\left(t_0\right)\right\rangle
\end{gathered}
$$

Thus, in the Heisenberg picture, the states \$\left|\psi\_H\left(t\_0\right)\right\rangle\$ do not change with time, but the operators associated with the observables do. Although fully equivalent to the Schr√∂dinger picture, Heisenberg's viewpoint is closer to our classical intuition. After all, in classical mechanics, we associate time-dependent position and momentum functions \$x(t), p(t)\$ with classically moving particles. In the Heisenberg picture of quantum mechanics, they become time-dependent operators \$x\_n(t)\$ and \$p\_1(t)\$.

---

### üîπ Time Evolution of Operators in the Heisenberg Picture

We now derive a differential equation for the time evolution of operators in the Heisenberg picture, starting from their definition:

$$
A_H(t)=U^{+}\left(t, t_0\right) A U\left(t, t_0\right)
$$

In order not to lose generality, we may assume that \$A\$ also contains some intrinsic time dependence (such as, for example, time dependence arising from the interaction with some external electric field). In this case, we get, after performing the time derivatives:


$$
\frac{d A_H}{d t} = \frac{d U^{+}\left(t, t_0\right)}{d t} A(t) U\left(t, t_0\right) + U^{+}\left(t, t_0\right) \frac{\partial A}{\partial t} U\left(t, t_0\right) + U^{+}\left(t, t_0\right) A \frac{d U\left(t, t_0\right)}{d t}
$$

From the operator form of the TDSE, we have:

$$
\begin{aligned}
& i \hbar \frac{d U}{d t} = H U \quad \Rightarrow \quad \frac{d U}{d t} = \frac{1}{i \hbar} H U \\
& \frac{d U^{+}}{d t} = -\frac{1}{i \hbar}(H U)^{+} = -\frac{1}{i \hbar} U^{+} H
\end{aligned}
$$

From this, we get:

$$
\frac{d A_H}{d t} = U^{+} \left( \frac{\partial A}{\partial t} \right) U + \frac{1}{i \hbar} \left[ U^{+} A H U - U^{+} H A U \right]
$$

By definition, we have:

$$
\frac{\partial A_H}{\partial t} := U^{+} \frac{\partial A}{\partial t} U
$$

and, after inserting the identity operator at appropriate places:

$$
\frac{d A_H}{d t} = \frac{\partial A_H}{\partial t} + \frac{1}{i \hbar} \left[ U^{+} A U\, U^{+} H U - U^{+} H U\, U^{+} A U \right]
$$

Since \$U^{+} H U = H\$ (why?) and \$A\_H = U^{+} A U\$, we obtain the equation of motion for Heisenberg operators:

$$
\begin{aligned}
\frac{d A_H}{d t} &= \frac{\partial A_H}{\partial t} + \frac{1}{i \hbar} \left[ A_H H - H A_H \right] \\
&= \frac{\partial A_H}{\partial t} + \frac{1}{i \hbar} \left[ A_H, H \right]
\end{aligned}
$$

For time-independent Schr√∂dinger operators such as \$x\$, \$p\$, etc., we get:

$$
i \hbar \frac{d A_H}{d t} = \left[ A_H, H \right]
$$

We see that the time evolution of operators is determined by their commutators with the Hamiltonian!

---

**Connection to classical mechanics:**
Consider a classical mechanical system evolving according to Hamilton's equations:






$$
\dot{x} = \frac{\partial H}{\partial p}, \quad \dot{p} = -\frac{\partial H}{\partial x}
$$

How does a classical observable \$A(x, p, t)\$ evolve in time along classical trajectories that are solutions of Hamilton's equations? Well, let us calculate the time derivative!

$$
\begin{aligned}
\frac{d A}{d t} &= \frac{\partial A}{\partial t} + \frac{\partial A}{\partial x} \cdot \dot{x} + \frac{\partial A}{\partial p} \cdot \dot{p} \\
&= \frac{\partial A}{\partial t} + \frac{\partial A}{\partial x} \cdot \frac{\partial H}{\partial p} - \frac{\partial A}{\partial p} \cdot \frac{\partial H}{\partial x}
\end{aligned}
$$

We now "invent" a bit of new notation and define what is called the **Poisson bracket** of two classical phase space functions:

$$
\{A, H\} := \frac{\partial A}{\partial x} \cdot \frac{\partial H}{\partial p} - \frac{\partial A}{\partial p} \cdot \frac{\partial H}{\partial x}
$$

With this notation, the equation for \$dA/dt\$ can be rewritten as:

$$
\frac{d A}{d t} = \frac{\partial A}{\partial t} + \{A, H\}
$$

This has an analogous form to the quantum relation:

$$
\frac{d A_H}{d t} = \frac{\partial A_H}{\partial t} + \frac{1}{i \hbar} \left[ A_H, H \right]
$$

if we identify the quantum commutator with the classical Poisson bracket:

$$
\frac{1}{i \hbar} \left[ A_H, H \right] \xrightarrow[\text{classical limit}]{} \{A, H\}
$$

This may seem like a bit of cheating, since after all we have only defined a suggestive notation for the Poisson bracket. However, the real point is that quantum commutators and Poisson brackets, defined in the above way, share an algebraic structure and mathematical properties. The following properties can be easily shown to be valid both for quantum commutators and for classical Poisson brackets:






## üîπ Properties shared by Commutators and Poisson Brackets

Let \$A\$, \$B\$, and \$C\$ be observables (operators in quantum mechanics or smooth functions on phase space in classical mechanics). The following algebraic properties are satisfied both by the **quantum commutator** $\[A, B] = AB - BA\$ and the **classical Poisson bracket** \${A, B}\$:

### 1. **Linearity**:

$$
[\alpha A + \beta B, C] = \alpha [A, C] + \beta [B, C]
$$

### 2. **Antisymmetry**:

$$
[A, B] = -[B, A]
$$

### 3. **Leibniz Rule (Derivation Property)**:

$$
[AB, C] = A[B, C] + [A, C]B
$$

(Another commonly seen form of the product rule is:)

$$
[A, BC] = [A, B]C + B[A, C]
$$

(You also wrote this expression, possibly a misstatement of the Leibniz rule:)

$$
[A, B, C] = B[A, C] + [A, B]C
$$

### 4. **Jacobi Identity**:

$$
[A, [B, C]] + [B, [C, A]] + [C, [A, B]] = 0
$$

> ‚úÖ **All these properties remain valid if we replace commutator brackets with Poisson brackets.**

## üîπ Lie Algebras and the Geometric Structure Behind the Bracket

For those who like to know more about the connections between the structure of quantum and classical mechanics to the concepts also used in differential geometry I summarize some useful terms for further self study...

### üí° **Definition (Lie Algebra)**:

A **Lie algebra** is a vector space $V$ over a field (e.g., $\mathbb{R}$ or $\mathbb{C}$), equipped with a bilinear operation $[\cdot, \cdot]: V \times V \to V$, called the **Lie bracket**, satisfying:

1. **Bilinearity**: $[aA + bB, C] = a[A, C] + b[B, C]$ for all scalars $a,b$
2. **Antisymmetry**: $[A, B] = -[B, A]$
3. **Jacobi identity**: $[A, [B, C]] + [B, [C, A]] + [C, [A, B]] = 0$

In both classical and quantum mechanics, the set of observables with their respective brackets (Poisson or commutator) forms a **Lie algebra**.

---

## üîπ Relation to Derivations and Lie Derivatives

### ‚ñ∂Ô∏è **Leibniz Rule ‚á® Derivations**:

The Leibniz rule reflects that the bracket defines a **derivation**, i.e., a linear operation \$D\$ satisfying:

$$
D(AB) = D(A)B + A D(B)
$$

This is true for both the **Poisson bracket** (where \${f, \cdot}\$ acts as a derivation) and the **quantum commutator** (where $[A, \cdot]$ acts as a derivation).

In this sense, each observable generates a **flow** or **infinitesimal transformation** on the space of observables, much like a vector field acting via a **Lie derivative** in differential geometry.

---

### ‚ñ∂Ô∏è **Lie Derivatives**:

In differential geometry, the **Lie derivative** $\mathcal{L}\_X$ with respect to a vector field $X$ is also a derivation on the algebra of smooth functions or tensor fields:

$$
\mathcal{L}_X(fg) = (\mathcal{L}_X f) g + f (\mathcal{L}_X g)
$$

In classical mechanics, each function $H$ defines a **Hamiltonian vector field** and the Poisson bracket \${A, H}\$ is the Lie derivative of $A$ along $X\_H\$:

$$
\{A, H\} = \mathcal{L}_{X_H} A
$$

Thus, Poisson brackets are directly related to **Lie derivatives** generated by Hamiltonian flows on phase space.

---

### ‚ñ∂Ô∏è **Commutators and Lie Algebras in Quantum Mechanics**:

In quantum mechanics, observables form a non-commutative algebra, and the **commutator** $[A, \cdot]$ acts as a derivation on this algebra. This is why time evolution in the Heisenberg picture,

$$
\frac{dA_H}{dt} = \frac{1}{i\hbar}[A_H, H]
$$

resembles the action of a Lie derivative: the Hamiltonian \$H\$ generates a flow on the algebra of observables through the commutator.
