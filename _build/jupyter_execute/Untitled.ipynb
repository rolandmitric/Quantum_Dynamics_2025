
# Time-Dependent Perturbation Theory

## Motivation

After we have learned what wavepackets are, we now want to answer the questions:

- **How can quantum wavepackets be created?**

Broadly speaking, you can create a quantum wavepacket in the same way you create classical waves ‚Äî **by perturbing the system**!

Thus, we will examine the influence of different kinds of perturbations on the time-evolution of quantum systems.
