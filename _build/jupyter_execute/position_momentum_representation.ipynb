## Position and Momentum Representation

Let us now consider quantum systems whose dynamics take place in real space, such that a position $x$ is an observable taking any real value. According to the rules of quantum mechanics, we associate to the position observable a Hermitian operator $\hat{x}$ such that

$$
\hat{x}^\dagger = \hat{x},
$$

i.e., $\hat{x}$ is self-adjoint.

Let $|\psi\rangle \in \mathcal{H}$ be a normalized state,

$$
\langle \psi | \psi \rangle = 1.
$$

To such a state we can assign its **average position** (expectation value of position) defined by

$$
\langle x \rangle = \langle \psi | \hat{x} | \psi \rangle.
$$

In principle, we can label any state $|\psi\rangle \in \mathcal{H}$ by the value of its expectation $\langle x \rangle$. Thus we may write:

$$
|\psi\rangle \rightarrow |\psi_{\langle x \rangle} \rangle.
$$

This notation is informal and not unique, since many different states can share the same expectation value $\langle x \rangle$, but it is suggestive of the position "center" of the state.

---

## Translation Operator and Its Generator

We now define a unitary operator called the **translation operator**, denoted by $\hat{T}(\Delta x)$, that shifts a quantum state in space by a distance $\Delta x$. Specifically, this operator satisfies:

$$
\hat{T}^\dagger(\Delta x) \hat{T}(\Delta x) = \hat{T}(\Delta x) \hat{T}^\dagger(\Delta x) = \mathbb{I},
$$

i.e., $\hat{T}(\Delta x)$ is **unitary** for all $\Delta x \in \mathbb{R}$. $\Delta x$ itself is a real parameter that labels each translation operator. 


### Shift of Expectation Value

To fix the form of $\hat{T}(\Delta x)$, we demand that it shifts the expectation value of position:

$$
\left\langle \psi_{\langle x \rangle} \right| \hat{T}^\dagger(\Delta x) \hat{x} \hat{T}(\Delta x) \left| \psi_{\langle x \rangle} \right\rangle = \langle x \rangle + \Delta x.
$$

This condition implies the operator identity:

$$
\hat{T}^\dagger(\Delta x) \hat{x} \hat{T}(\Delta x) = \hat{x} + \Delta x \cdot \mathbb{I}.
$$

Multiplying both sides from the left with $\hat{T}(\Delta x)$, we obtain the **commutator relation**:

$$
\hat{x} \hat{T}(\Delta x) - \hat{T}(\Delta x) \hat{x} = \Delta x \cdot \hat{T}(\Delta x),
$$

or

$$
[\hat{x}, \hat{T}(\Delta x)] = \Delta x \cdot \hat{T}(\Delta x).
$$

We observe that the translation operator does not commute with the position operator. 


## Momentum as Generator of Translations

To explore the structure of the translation operator, we differentiate the unitarity condition:

$$
\hat{T}^\dagger(x) \hat{T}(x) = \mathbb{I}.
$$

Differentiating both sides with respect to $x$, we obtain

$$
\frac{\partial \hat{T}}{\partial x} \hat{T}^\dagger + \hat{T} \frac{\partial \hat{T}^\dagger}{\partial x} = 0.
$$

This implies that

$$
\left( \frac{\partial \hat{T}}{\partial x} \hat{T}^\dagger \right) + \left( \frac{\partial \hat{T}}{\partial x} \hat{T}^\dagger \right)^\dagger = 0,
$$

which shows that the operator

$$
\frac{\partial \hat{T}}{\partial x} \hat{T}^\dagger
$$

is **anti-Hermitian**.

We denote this anti-Hermitian operator by $\pi$, and write

$$
\frac{\partial \hat{T}}{\partial x} \hat{T}^\dagger = \pi.
$$

Multiplying both sides on the right by $\hat{T}$, we get the differential equation

$$
\frac{\partial \hat{T}}{\partial x} = \pi \hat{T}.
$$

Assuming $\pi$ is independent of $x$, this can be formally solved as

$$
\hat{T}(x) = e^{\pi x}.
$$

We now **define the momentum operator $\hat{p}$** as the Hermitian generator of translations by setting

$$
\pi = -\frac{i}{\hbar} \hat{p}.
$$

Then the translation operator becomes

$$
\hat{T}(x) = e^{- \frac{i}{\hbar} \hat{p} x}.
$$

For $x = \Delta x$, we thus have

$$
\hat{T}(\Delta x) = e^{- \frac{i}{\hbar} \hat{p} \Delta x}.
$$

---

### Translation as a One-Parameter Unitary Group

From a mathematical point of view, the family of translation operators $\{ \hat{T}(x) \}_{x \in \mathbb{R}}$ forms a **one-parameter unitary group** under composition as a group operation:

1. $\hat{T}(0) = \mathbb{I}$ (identity element),
2. $\hat{T}(x_1)\hat{T}(x_2) = \hat{T}(x_1 + x_2)$ (group property),
3. $\hat{T}(x)$ is continuous in the parameter $x \in \mathbb{R}$.

These properties imply that translations in space form a continuous group structure where each group element is labeled by a real parameter $x$, and multiplication of translation operators corresponds to addition of their displacement parameters.

### Connection to Stone’s Theorem

According to **Stone’s theorem** on one-parameter unitary groups, every (strongly) continuous one-parameter unitary group $\{ U(t) \}_{t \in \mathbb{R}}$ on a Hilbert space $\mathcal{H}$ can be written in the form

$$
U(t) = e^{-i t A}
$$

for some unique **self-adjoint operator** $A$, which is called the **generator** of the unitary group.

In our case, the group $\hat{T}(x)$ is parametrized by the real displacement $x$, and the generator of spatial translations is a Hermitian operator which we identify with the **momentum operator**:

$$
\hat{T}(x) = e^{- \frac{i}{\hbar} \hat{p} x}.
$$

This shows that the momentum operator $\hat{p}$ is (up to a factor $\hbar$) the generator of spatial translations in quantum mechanics meaning that the infinitesimal translations are induced by the action of $\hat{p}$ on a given state.


## Commutator of $\hat{x}$ and $\hat{p}$

We now ask the question whether the position and the momentum as we have defined them are simultaeously determined. In order to answer this we must find out wheter the position and momentum operators commute. 
We recall the earlier derived commutator:

$$
[\hat{x}, \hat{T}( x)] =  x \cdot \hat{T}(\Delta x).
$$

Inserting the exponential form of $\hat{T}$, we get

$$
\left[\hat{x}, e^{- \frac{i}{\hbar} \hat{p} x} \right] = x \cdot e^{- \frac{i}{\hbar} \hat{p} x}.
$$

Differentiating both sides with respect to $x$, we get

$$
\left[\hat{x}, -\frac{i}{\hbar} \hat{p} e^{- \frac{i}{\hbar} \hat{p} x} \right] = e^{- \frac{i}{\hbar} \hat{p} x} + x \left(-\frac{i}{\hbar} \hat{p}\right) e^{- \frac{i}{\hbar} \hat{p} x}.
$$

Now set $x = 0$. Using the fact that $e^0 = \mathbb{I}$, we find

$$
\left[\hat{x}, -\frac{i}{\hbar} \hat{p} \right] = \mathbb{I}
\quad \Rightarrow \quad
[\hat{x}, \hat{p}] = i \hbar.
$$

Thus, **position and momentum do not commute**. Their canonical commutation relation is:

$$
[\hat{x}, \hat{p}] = i \hbar.
$$

This is one of the fundamental structures of quantum mechanics and leads directly to  the Heisenberg's uncertainty principle.

---

## Eigenstates of Position and Momentum

The fact that $\hat{x}$ and $\hat{p}$ do not commute implies that there is **no basis** in which both operators are diagonal—there are **no common eigenstates**.

Assume $\hat{x}$ has eigenstates $|x\rangle$ satisfying:

$$
\hat{x} |x\rangle = x \cdot |x\rangle.
$$

We assume these states form a **complete set** in the Hilbert space:

$$
\int dx\, |x\rangle\langle x| = \mathbb{I}.
$$

Then any state $|\psi\rangle$ can be decomposed as:

$$
|\psi\rangle = \int dx' \, |x'\rangle \langle x' | \psi \rangle.
$$

Define the **position-space wavefunction** as

$$
\psi(x') := \langle x' | \psi \rangle.
$$

Then projecting onto $\langle x|$ gives:

$$
\langle x | \psi \rangle = \int dx' \langle x | x' \rangle \psi(x').
$$

This shows that $\langle x | x' \rangle$ acts like a **Dirac delta function**:

$$
\langle x | x' \rangle = \delta(x - x'), \quad \text{with} \quad \int dx' \delta(x - x') = 1.
$$

---

## Action of Operators in Position Representation

Let us now examine how operators act in the position representation.

### Position operator

We compute:

$$
\langle x | \hat{x} | \psi \rangle = x \cdot \langle x | \psi \rangle = x \cdot \psi(x).
$$

Thus:

$$
\hat{x} \psi(x) = x \cdot \psi(x).
$$

This means that in the position representation, the position operator acts as **multiplication by $x$**.

---

### Momentum operator

We now compute how $\hat{p}$ acts on a wavefunction $\psi(x)$. Recall that

$$
\hat{T}(\Delta x) = e^{- \frac{i}{\hbar} \hat{p} \Delta x}
\approx \mathbb{I} - \frac{i}{\hbar} \hat{p} \Delta x + \mathcal{O}(\Delta x^2),
$$

so that:

$$
\hat{p} = i \hbar \lim_{\Delta x \to 0} \frac{\hat{T}(\Delta x) - \mathbb{I}}{\Delta x}.
$$

Then:

$$
\begin{aligned}
\langle x | \hat{p} | \psi \rangle
&= i \hbar \lim_{\Delta x \to 0} \frac{1}{\Delta x} \left( \langle x | \hat{T}(\Delta x) | \psi \rangle - \langle x | \psi \rangle \right) \\
&= i \hbar \lim_{\Delta x \to 0} \frac{1}{\Delta x} \left( \psi(x - \Delta x) - \psi(x) \right) \\
&= -i \hbar \frac{d\psi}{dx}(x).
\end{aligned}
$$

Hence:

$$
\hat{p} \psi(x) = -i \hbar \frac{d}{dx} \psi(x).
$$

So, in the position representation, the momentum operator acts as a **differential operator**:

$$
\hat{p} = -i \hbar \frac{d}{dx}.
$$




##  Heisenberg’s Uncertainty Relation


We now examine to what extent the position and the momentum of a quantum particle can be simultaneously determined.

Let us define the uncertainty in the position and in the momentum as the expectation value of the squared deviation from the mean:

$$
\begin{aligned}
& \left\langle(\Delta x)^2\right\rangle := \langle \psi | (\hat{x} - \langle x \rangle)^2 | \psi \rangle \\
& \left\langle(\Delta p)^2\right\rangle := \langle \psi | (\hat{p} - \langle p \rangle)^2 | \psi \rangle
\end{aligned}
$$

We are interested in the product of the position and momentum variances:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle = 
\langle \psi | (\hat{x} - \langle x \rangle)^2 | \psi \rangle \cdot \langle \psi | (\hat{p} - \langle p \rangle)^2 | \psi \rangle.
$$

Since both \( \hat{x} - \langle x \rangle \) and \( \hat{p} - \langle p \rangle \) are self-adjoint operators, we can write these variances as squared norms of the corresponding vectors:

$$
\begin{aligned}
\left\langle(\Delta x)^2\right\rangle &= \| (\hat{x} - \langle x \rangle) \psi \|^2 \\
\left\langle(\Delta p)^2\right\rangle &= \| (\hat{p} - \langle p \rangle) \psi \|^2
\end{aligned}
$$

Their product is therefore:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle 
= \| (\hat{x} - \langle x \rangle) \psi \|^2 \cdot \| (\hat{p} - \langle p \rangle) \psi \|^2.
$$

We now apply the **Cauchy-Schwarz inequality**:

$$
| \langle a | b \rangle |^2 \leq \|a\|^2 \cdot \|b\|^2,
$$

with \( a = (\hat{x} - \langle x \rangle)\psi \) and \( b = (\hat{p} - \langle p \rangle)\psi \). This yields:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle 
\geq \left| \langle (\hat{x} - \langle x \rangle)\psi \mid (\hat{p} - \langle p \rangle)\psi \rangle \right|^2.
$$

Let us denote this inner product as:

$$
\langle (\Delta x)\psi \mid (\Delta p)\psi \rangle := \langle \psi | (\hat{x} - \langle x \rangle)(\hat{p} - \langle p \rangle) | \psi \rangle.
$$

We now separate this complex number into real and imaginary parts. Let:

$$
\langle A \rangle := \langle \psi | (\hat{x} - \langle x \rangle)(\hat{p} - \langle p \rangle) | \psi \rangle.
$$

Then:

$$
\begin{aligned}
\left| \langle A \rangle \right|^2 &= \left( \operatorname{Re} \langle A \rangle \right)^2 + \left( \operatorname{Im} \langle A \rangle \right)^2 \\
&\geq \left( \operatorname{Im} \langle A \rangle \right)^2.
\end{aligned}
$$

We now evaluate the imaginary part:

$$
\begin{aligned}
\operatorname{Im} \langle A \rangle 
&= \frac{1}{2i} \left( \langle A \rangle - \langle A \rangle^* \right) \\
&= \frac{1}{2i} \left( \langle \psi | (\hat{x} - \langle x \rangle)(\hat{p} - \langle p \rangle) | \psi \rangle 
- \langle \psi | (\hat{p} - \langle p \rangle)(\hat{x} - \langle x \rangle) | \psi \rangle \right) \\
&= \frac{1}{2i} \langle \psi | [\hat{x}, \hat{p}] | \psi \rangle = \frac{1}{2i} \cdot i\hbar = \frac{\hbar}{2}.
\end{aligned}
$$

Thus:

$$
\left| \langle A \rangle \right|^2 \geq \left( \frac{\hbar}{2} \right)^2.
$$

Putting everything together:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle \geq \left( \frac{\hbar}{2} \right)^2,
$$

and finally,

$$
\sqrt{\left\langle(\Delta x)^2\right\rangle} \cdot \sqrt{\left\langle(\Delta p)^2\right\rangle} \geq \frac{\hbar}{2}.
$$

This is the **Heisenberg uncertainty relation**.

### Minimum uncertainty wave packet and derivation of the Fourier kernel

We have seen that there is a lower bound to the product of the uncerntainties in the position and momentum. We now construct special kind of states for which the product of the uncerntainties reaches its minimal value. 

A necessary and sufficient condition for such a minimum uncertainty state is that it satisfies the operator identity:

$$
(\hat{p} - \langle \hat{p} \rangle)|\psi\rangle = \lambda (\hat{x} - \langle \hat{x} \rangle)|\psi\rangle,
$$

where $ \lambda \in \mathbb{C} $. We now derive the form of such a wavefunction in both position and momentum representation.

---

#### Solution in position representation

Projecting onto position eigenstates and using $ \hat{p} = -i\hbar \frac{d}{dx} $, $ \hat{x} = x $, we obtain a first-order differential equation for the position representation of the minimum uncertainty state:

$$
\left( -i\hbar \frac{d}{dx} - \langle p \rangle \right)\psi(x) = \lambda (x - \langle x \rangle) \psi(x).
$$

Since the variables are separated we can rewrite:

$$
\frac{d\psi}{dx} = \frac{i}{\hbar} \left[ \lambda(x - \langle x \rangle) + \langle p \rangle \right] \psi(x).
$$

and integrate:

$$
\psi(x) = A \exp\left[ \frac{i\lambda}{2\hbar}(x - \langle x \rangle)^2 + \frac{i\langle p \rangle}{\hbar} x \right].
$$

To ensure normalizability, we must set $ \lambda = i a $ to be purely imaginary, with $ a > 0 $. Using these properties we can define $ a = \hbar / (2\sigma_x^2) $. Then the state becomes:

$$
\psi(x) = N_x \exp\left[ -\frac{(x - \langle x \rangle)^2}{4\sigma_x^2} + \frac{i\langle p \rangle}{\hbar} x \right],
$$

with normalization constant given by:

$$
N_x = \left( \frac{1}{2\pi \sigma_x^2} \right)^{1/4}.
$$

---

#### Solution in momentum representation


The same problem can of course be also solved in the momentum basis. Recalling how the operators act in the momentum basis:  $ \hat{p} = p\times $, $ \hat{x} = i\hbar \frac{d}{dp} $, the same condition becomes:

$$
(p - \langle p \rangle)\phi(p) = i\hbar \lambda \frac{d\phi}{dp} - \lambda \langle x \rangle \phi(p),
$$

which simplifies again to a first-order differential equation that now reads:

$$
\frac{d\phi}{dp} = \frac{1}{i\hbar \lambda} \left[ (p - \langle p \rangle) + \lambda \langle x \rangle \right] \phi(p).
$$

Integration gives:

$$
\phi(p) = A' \exp\left[ \frac{1}{2 i\hbar \lambda}(p - \langle p \rangle)^2 + \frac{\langle x \rangle}{i\hbar} p \right].
$$

Setting again $ \lambda = i a $, with $ \sigma_p = \hbar / (2\sigma_x)$ we obtain:

$$
\phi(p) = N_p \exp\left[ -\frac{(p - \langle p \rangle)^2}{4\sigma_p^2} + \frac{i \langle x \rangle}{\hbar} p \right],
$$

with

$$
N_p = \left( \frac{1}{2\pi \sigma_p^2} \right)^{1/4}.
$$


Since $\psi(x)$ and $\phi(p)$ are two representations of the same state in two different bases they must be related by a linear transformation. In other words, the position-space wave function $\psi(x)$ can be expressed as a continuous superposition of momentum eigenstates (and vice versa!) by:

$$
\psi(x) = \int_{-\infty}^{+\infty} K(x, p)\, \phi(p)\, dp,
$$

where $\phi(p)$ is the wave function in the momentum representation, and $K(x, p)$ is the transformation kernel from momentum to position basis. This kernel must implement a **unitary transformation** from one basis to another which is a continuous analogue of a change of orthonormal basis in finite-dimensional Hilbert spaces.

We now assume that the momentum-space wave function $\phi(p)$ is given by the **minimum uncertainty wave packet** (centered at $\langle p \rangle$, with width $\sigma_p$):

$$
\phi(p) = N_p \cdot \exp\left[ -\frac{(p - \langle p \rangle)^2}{4\sigma_p^2} + \frac{i}{\hbar} \langle x \rangle p \right],
$$

where $N_p$ is a normalization constant.

Substituting this into the expression for $\psi(x)$, we obtain:

$$
\psi(x) = N_p \int_{-\infty}^{+\infty} K(x, p)\, \exp\left[ -\frac{(p - \langle p \rangle)^2}{4\sigma_p^2} + \frac{i}{\hbar} \langle x \rangle p \right] dp.
$$


Since the position representation of the state is also a Gaussian, and remembering the properties of Gaussian integrals that we used before, we realize that in order for the relation

$$
\psi(x) = \int_{-\infty}^{+\infty} K(x, p)\, \phi(p)\, dp
$$

to yield a Gaussian in $x$, the kernel must contain an exponential factor of the form

$$
K(x, p) = \mathcal{N} \cdot e^{\frac{i}{\hbar} x p},
$$

where $\mathcal{N}$ is a normalization constant that needs to be properly chosen.

Substituting this into the integral representation of the wave function, using the known form of the minimum uncertainty wave packet in momentum space,

$$
\phi(p) = N_p \cdot \exp\left[ -\frac{(p - \langle p \rangle)^2}{4\sigma_p^2} + \frac{i}{\hbar} \langle x \rangle p \right],
$$

we obtain

$$
\psi(x) = \mathcal{N} N_p \int_{-\infty}^{+\infty} \exp\left[ -\frac{(p - \langle p \rangle)^2}{4\sigma_p^2} + \frac{i}{\hbar} \langle x \rangle p + \frac{i}{\hbar} x p \right] dp.
$$

This simplifies to

$$
\psi(x) = \mathcal{N} N_p \int_{-\infty}^{+\infty} \exp\left[ -\frac{(p - \langle p \rangle)^2}{4\sigma_p^2} + \frac{i}{\hbar}(x + \langle x \rangle) p \right] dp.
$$

This is a **Gaussian integral** in $p$, and hence can be evaluated analytically. The exponent is a quadratic polynomial in $p$, so the result of the integration is again a Gaussian — now in $x$.

Carrying out the integration (and absorbing constant factors into $N_x$), we obtain the known result

$$
\psi(x) = N_x \cdot \exp\left[ -\frac{(x - \langle x \rangle)^2}{4\sigma_x^2} + \frac{i}{\hbar} \langle p \rangle x \right],
$$

with $\sigma_x \sigma_p = \hbar / 2$.

This confirms that the transformation kernel

$$
\boxed{
K(x, p) = \frac{1}{\sqrt{2\pi \hbar}} \cdot e^{\frac{i}{\hbar} x p}
}
$$

maps the Gaussian wave packet in momentum space to its correct position space counterpart. This is the **Fourier kernel**, now derived directly from the requirement that it preserves the form of the minimum uncertainty wave packet.

This is the **Fourier transformation kernel** that relates the momentum and position representations in quantum mechanics. We have derived it purely from the requirement that it transforms the minimum uncertainty wave packet $\phi(p)$ into its correct position-space counterpart $\psi(x)$.
