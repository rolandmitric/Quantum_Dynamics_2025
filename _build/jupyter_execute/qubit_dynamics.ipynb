## ðŸ”¸ Quantum Dynamics of a Two-Level System with Constant Coupling

As a first physical application of the formalism we have developed â€” Hilbert spaces, linear operators, and time evolution â€” we now turn to a **two-level quantum system**.
This model is often used to illustrate various features of quantum mechanics and it also has significant practical relevance. Spin-1/2 particles such as electrons and protons can be treated as two level systems in other degrees of freedom can be ignored as is the case in the nuclear or magnetic resonance spectroscopies. Also the basic unit of the quantum information processing - the qubit - is a two level quantum system that can be realized in various physical ways. Two-level systems are also used in the field of chemical physics as a model for molecular systems where only two electronic states couple or as model describing the interaction of trapped atomic ions with laser fields. 

### ðŸ”¹ Basis States and Time-Dependent Wavefunction

We work in the orthonormal basis of diabatic (or bare) states:

- $ |g\rangle $ : ground state
- $ |e\rangle $ : excited state

The general state of the system at time $t$ is written as a time-dependent linear combinations of the basis states:

$$
|\psi(t)\rangle = c_g(t) |g\rangle + c_e(t) |e\rangle
$$

where $c_g(t), c_e(t) \in \mathbb{C} $ are the **time-dependent amplitudes**.

---

### ðŸ”¹ 1. Hamiltonian with Constant Coupling

The first system we treat is the one with a time-independent Hamiltonian of the form:

$$
\hat{H} = 
\begin{pmatrix}
E_g & V \\
V & E_e
\end{pmatrix}
$$

where:

- $ E_g $ and $ E_e $ are the energies of the ground and excited states, respectively,
- $ V \in \mathbb{R} $ is the **constant coupling** between the two states.


We will analyze the dynamics in two regimes:

1. **Resonant Case**: $E_g = E_e$  
   The two levels are degenerate. As we will demonstrate, the constant coupling leads in this situtation to **maximal population transfer** and perfectly symmetric oscillations.

2. **Non-Resonant Case**: $E_g \ne E_e$  
   The energy levels are detuned. While the population transfer still occurs, the oscillations are asymmetric, and their amplitude depends on the energy difference.

Our goal is in both cases to compute the **time-evolution operator** explicitly and investigate the behavior of the time-dependent state populations that are defined by:

$$
P_g(t) = |\langle g | \psi(t)\rangle|^2, \quad P_e(t) = |\langle e | \psi(t)\rangle|^2
$$




## ðŸ”¸ Two-Level System: Dynamics via Analytic and Symbolic Methods

Continuing our study of a two-level system with constant coupling, we now explicitly evaluate the **time-evolution operator**.

We proceed in two complementary ways:

1. **Analytic derivation**: Using matrix exponentials.
2. **Symbolic computation**: Using the SymPy package.


The Hamiltonian of any two-level system can be represented by a 2x2 hermitian matrix in the basis $|g\rangle , |e\rangle$. Such matrices can be expanded in terms of the 
 **Pauli matrices** and the **identity matrix**:

$$
\sigma_x =
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}, \quad
\sigma_y =
\begin{pmatrix}
0 & -i \\
i & 0
\end{pmatrix}, \quad
\sigma_z =
\begin{pmatrix}
1 & 0 \\
0 & -1
\end{pmatrix}, \quad
I =
\begin{pmatrix}
1 & 0 \\
0 & 1
\end{pmatrix}
$$

In this basis the Hamiltonian of any two level system can be written as: 

$$
\hat{H} = \overline{E} I + V_{x} \sigma_x + V_{y} \sigma_y + \frac{1}{2}\Delta \sigma_z
$$

where:

- $\overline{E} \in \mathbb{R}$ is the average energy of the two levels,
- $\Delta \in \mathbb{R}$ is the energy difference between the ground and excited states,
- $V_{x}$, $V_{y} \in \mathbb{R}$ are real and imaginary parts of the coupling.

---

## ðŸ”¹ Special Case: Rabi Oscillations (Resonant Coupling)

We specialize to the situation:

- $\overline{E} = 0$,
- $\Delta = 0$,
- $V_{y} = 0$,
- $V_{x} = V$,


so that the Hamiltonian simplifies to:

$$
\hat{H} = V \sigma_x
$$

i.e.,

$$
\hat{H} =
V
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
$$


## ðŸ”¹ 1. Analytic Derivation of the Propagator

In the following we calculate the time-evolution operator for the two-level system Hamiltonian analytically using the exponential for for the time-evolution operator:
The Hamiltonian in our case is given by:

$$
\hat{H} = V \sigma_x
$$

which leads to the following time-evolution operator:

$$
\hat{U}(t) = \exp\left( -\frac{i}{\hbar} \hat{H} t \right) = \exp\left( -i \frac{V}{\hbar} t \sigma_x \right)
$$

In order to calculate the operator exponential we need to know what the powers of the $\sigma_x$ are.
Recalling properties of the Pauli matrices we have:

$$
\sigma_x^2 = I
$$

where $I$ is the $2\times 2$ identity matrix.

Higher powers of $\sigma_x$ cycle periodically between two values:

$$
\begin{aligned}
\sigma_x^0 &= I, \\
\sigma_x^1 &= \sigma_x, \\
\sigma_x^2 &= I, \\
\sigma_x^3 &= \sigma_x, \\
\sigma_x^4 &= I, \quad \text{and so on.}
\end{aligned}
$$

Thus, for any integer $n$, we have generally:

$$
\sigma_x^{2n} = I, \quad \sigma_x^{2n+1} = \sigma_x
$$


Now we can calculate time-evolution operator using the Taylor series expansion of the exponential function:

$$
e^A = \sum_{n=0}^{\infty} \frac{A^n}{n!}
$$

Applying this to $A = -i \frac{V t}{\hbar} \sigma_x$ we obtain:

$$
\begin{aligned}
\hat{U}(t)
&= \sum_{n=0}^{\infty} \frac{1}{n!} \left( -i \frac{V t}{\hbar} \sigma_x \right)^n \\
&= \sum_{n=0}^{\infty} \frac{1}{n!} \left( -i \frac{V t}{\hbar} \right)^n \sigma_x^n
\end{aligned}
$$

We can now split the sum into **even** and **odd** terms:

$$
\begin{aligned}
\hat{U}(t)
&= \sum_{k=0}^{\infty} \frac{1}{(2k)!} \left( -i \frac{V t}{\hbar} \right)^{2k} \sigma_x^{2k}
+ \sum_{k=0}^{\infty} \frac{1}{(2k+1)!} \left( -i \frac{V t}{\hbar} \right)^{2k+1} \sigma_x^{2k+1} \\
&= \left[ \sum_{k=0}^{\infty} \frac{1}{(2k)!} \left( -i \frac{V t}{\hbar} \right)^{2k} \right] I
+ \left[ \sum_{k=0}^{\infty} \frac{1}{(2k+1)!} \left( -i \frac{V t}{\hbar} \right)^{2k+1} \right] \sigma_x
\end{aligned}
$$

A well trained eye will recognize in these ugly expressions the series definining the sine and cosine functions:

$$
\cos\left( \frac{V t}{\hbar} \right) = \sum_{k=0}^{\infty} \frac{(-1)^k}{(2k)!} \left( \frac{V t}{\hbar} \right)^{2k}
$$


$$
\sin\left( \frac{V t}{\hbar} \right) = \sum_{k=0}^{\infty} \frac{(-1)^k}{(2k+1)!} \left( \frac{V t}{\hbar} \right)^{2k+1}
$$

Furthermore, we note that:

$
\left( -i \right)^{2k} = (-1)^k, \quad \left( -i \right)^{2k+1} = (-1)^k (-i)
$

Thus:

- For even powers:

$$
\left( -i \right)^{2k} \left( \frac{V t}{\hbar} \right)^{2k} = (-1)^k \left( \frac{V t}{\hbar} \right)^{2k}
$$

- For odd powers:

$$
\left( -i \right)^{2k+1} \left( \frac{V t}{\hbar} \right)^{2k+1} = (-1)^k (-i) \left( \frac{V t}{\hbar} \right)^{2k+1}
$$


Collecting these results we get the time-evolution operator in a closed analytic form:

$$
\hat{U}(t) =
\cos\left( \frac{V t}{\hbar} \right) I
- i \sin\left( \frac{V t}{\hbar} \right) \sigma_x
$$

that can be also written explicitly as a 2x2 matrix operator:

$$
\hat{U}(t) =
\begin{pmatrix}
\cos\left( \frac{V t}{\hbar} \right) & -i \sin\left( \frac{V t}{\hbar} \right) \\
-i \sin\left( \frac{V t}{\hbar} \right) & \cos\left( \frac{V t}{\hbar} \right)
\end{pmatrix}
$$

---

## ðŸ”¹ 2. Symbolic Derivation Using SymPy for those who like symbolic computations...

For those familiar with python I also provide a short symbolic derivation of the above result using **SymPy**.

```python
import sympy as sp

# Define Pauli matrices and identity
sigma_x = sp.Matrix([[0, 1], [1, 0]])
sigma_y = sp.Matrix([[0, -sp.I], [sp.I, 0]])
sigma_z = sp.Matrix([[1, 0], [0, -1]])
Id = sp.eye(2)

# Define symbols
H0, Delta, V_re, V_im = sp.symbols('H0 Delta V_re V_im', real=True)
t, hbar = sp.symbols('t hbar', real=True, positive=True)

# General Hamiltonian
H = H0 * Id + V_re * sigma_x + V_im * sigma_y + (Delta/2) * sigma_z

# Specialize to resonant Rabi case
H = H.subs({H0: 0, Delta:0, V_im: 0})
H = sp.simplify(H)
# Compute propagator
Ht = -sp.I * H * t / hbar
U = sp.simplify(Ht.exp())  # Matrix exponential
U
```

## ðŸ”¹ Population Dynamics

We now study the dynamics assuming that the system starts in the  ground state $|\psi(0)\rangle = |g\rangle$ at the initial time $t=0$:
The ground state population as the system evolves can be calculated by applying the time-evolution operator and calculating the absolute square of the population which leads to:

$$
P_e(t) =  \sin^2\left( \frac{V t}{\hbar} \right)
$$

For the ground state we get:

$$
P_g(t) = \cos^2\left( \frac{V t}{\hbar} \right)
$$

As it should be the total population is normalized $P_g(t) + P_e(t) = 1$. The plot of the population dynamics is provided in the Figure below.

---

```python
import numpy as np
import matplotlib.pyplot as plt

hbar = 1.0  
V_re = 1.0 
t = np.linspace(0, 5 * np.pi, 1000)  # From 0 to 10 * pi
P_g = np.cos(V_re * t / hbar)**2
P_e = np.sin(V_re * t / hbar)**2

plt.figure(figsize=(7, 5))
plt.plot(t, P_g, color = "r", label=r'$P_g(t) = \cos^2\left(\frac{V_{\mathrm{re}} t}{\hbar}\right)$', linewidth=2)
plt.plot(t, P_e, color = "b", label=r'$P_e(t) = \sin^2\left(\frac{V_{\mathrm{re}} t}{\hbar}\right)$', linestyle='--', linewidth=2)
plt.title('Rabi Oscillations in a Two-Level System', fontsize=16)
plt.xlabel('Time $t$', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ðŸ”¸ Ï€/2 Pulse and the Quantum NOT Gate

We now consider a central concept in both quantum dynamics and quantum information theory: the **Ï€/2 pulse**.

A Ï€/2 pulse refers to a resonant driving of a two-level system such that the system undergoes a **coherent population transfer** from the ground state $|g\rangle$ to the excited state $|e\rangle$. This is a key example of **Rabi oscillations**.

### ðŸ”¹ Dynamics under a Ï€/2 Pulse

Let us return to the resonant Hamiltonian:

$$
\hat{H} = V \sigma_x = V
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
$$

The time-evolution operator is:

$$
\hat{U}(t) = \exp\left(-\frac{i}{\hbar} \hat{H} t \right) = \cos\left(\frac{Vt}{\hbar}\right) I - i \sin\left(\frac{Vt}{\hbar}\right) \sigma_x
$$

Acting on an initial ground state $|\psi(0)\rangle = |g\rangle = \begin{pmatrix}1 \\ 0\end{pmatrix}$, we find the time-evolved state:

$$
|\psi(t)\rangle = \cos\left(\frac{Vt}{\hbar}\right)|g\rangle - i \sin\left(\frac{Vt}{\hbar}\right)|e\rangle
$$

For a **Ï€/2 pulse**, we choose the interaction time $t$ such that:

$$
\frac{Vt}{\hbar} = \frac{\pi}{2}
\quad \Rightarrow \quad
t = \frac{\pi \hbar}{2V}
$$

At this moment, the system is found in:

$$
|\psi(t)\rangle = -i |e\rangle
$$

up to a global phase. Hence, the system has **fully transferred** from $|g\rangle$ to $|e\rangle$.

---

### ðŸ”¹ Connection to the Quantum NOT Gate

The **quantum NOT gate** (also called the Pauli-X gate) is defined by its action:

$$
X |0\rangle = |1\rangle, \quad X |1\rangle = |0\rangle
$$

This gate is represented by the same matrix as $\sigma_x$:

$$
X =
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
$$

In the context of Rabi dynamics, a Ï€ pulse (not Ï€/2) maps $|g\rangle \rightarrow |e\rangle$ and $|e\rangle \rightarrow |g\rangle$, just as the NOT gate does. Therefore, the Ï€ pulse **implements a quantum NOT gate** in the two-level system. The Ï€/2 pulse, on the other hand, is associated with a **Hadamard-like** transformation (placing the system into a coherent superposition), but a Ï€ pulse is the true NOT operation.

---

### ðŸ”¹ Physical Realizations

Quantum NOT gate described above is a simple example of a quantum gate that have been by now realized in a variety of platforms used to implement the qubits:

- **Trapped Ions**: Quantum operations are  implemented by resonant laser pulses that drive transitions between internal electronic levels of ions trapped in electromagnetic fields.
  
  â†’ *Cirac, J. I., & Zoller, P. (1995). Quantum computations with cold trapped ions. Physical Review Letters, 74(20), 4091.*

- **Superconducting Qubits**: Microwave pulses play the role of driving fields for Josephson junction-based qubits, performing quantum gates by engineering the duration and phase of the pulse.

  â†’ *Barends, R., et al. (2014). Superconducting quantum circuits at the surface code threshold for fault tolerance. Nature, 508(7497), 500â€“503.*

- **NMR Qubits**: Nuclear magnetic resonance experiments naturally implement single-qubit gates as rotations driven by radiofrequency fields.

  â†’ *Cory, D. G., Fahmy, A. F., & Havel, T. F. (1997). Ensemble quantum computing by NMR spectroscopy. PNAS, 94(5), 1634â€“1639.*

---

This illustrates how **control over interaction time** (via the pulse area) allows direct implementation of logical operations on qubits in physical systems.


## ðŸ”¸ Two-Level System: Non-Resonant Coupling (Detuned Case)

We now extend the analysis to a situation where the two states are **not degenerate**. The finite energy difference between the ground and excited states is denoted by $\Delta$.
We will see that the fact that the states are not resonant leads to **detuned Rabi oscillations** with reduced amplitude and also a reduced frequencies that are dependent on the detuning.

---

## ðŸ”¹ Hamiltonian with Detuning

The two-level Hamiltonian now has an aditional term that is proportional to the Pauli z-Matrix:

$$
\hat{H} = V_{\text{re}} \sigma_x + \Delta \sigma_z.
$$

Here we use the followung notation:

- $V_{\text{re}}$ is the real coupling strength,
- $\Delta$ is the energy splitting between the two bare states.

Written explictly the Hamiltonian matrix thus reads:

$$
\hat{H} =
\begin{pmatrix}
\Delta & V_{\text{re}} \\
V_{\text{re}} & -\Delta
\end{pmatrix}
$$

---

## ðŸ”¹ Algebraic Derivation of the Time Evolution Operator


Again, we calculate the **time-evolution operator** using algebraic properties of the Pauli-matrices starting from the exponential form of the operator:

$$
\hat{U}(t) = \exp\left( -\frac{i}{\hbar} \hat{H} t \right)
$$

To compute $\hat{U}(t)$, we expand the exponential using the Taylor series:

$$
\hat{U}(t) = \sum_{n=0}^{\infty} \frac{1}{n!} \left( -\frac{i}{\hbar} t \hat{H} \right)^n
$$

Thus, in order to construct the time-evolution operators we need to calculate the powers of $\hat{H}$ which now requires calculating powers of somewhat more complicated combination of Pauli-matrices. In order to proceed we will use the following relations between the Pauli-matrices:

$$
\begin{aligned}
\sigma_x^2 &= I, \quad \sigma_z^2 = I, \\
\sigma_x \sigma_z &= i \sigma_y, \quad \sigma_z \sigma_x = -i \sigma_y, \\
\sigma_x \sigma_z + \sigma_z \sigma_x &= 0 \quad \text{(anticommutation relation)}.
\end{aligned}
$$

We start by calculating the second power of the Hamiltonian and subsequently generalize inductively to higher powers:

$$
\begin{aligned}
\hat{H}^2
&= (V_{\text{re}} \sigma_x + \Delta \sigma_z)^2 \\
&= V_{\text{re}}^2 \sigma_x^2 + \Delta^2 \sigma_z^2 + V_{\text{re}} \Delta (\sigma_x \sigma_z + \sigma_z \sigma_x) \\
&= (V_{\text{re}}^2 + \Delta^2) I
\end{aligned}
$$

Thus:

$$
\hat{H}^2 = (V_{\text{re}}^2 + \Delta^2) I
$$

This can now be straightforwardly generalized to higher powers leading to:

- $\hat{H}^{2n} = (V_{\text{re}}^2 + \Delta^2)^n I$
- $\hat{H}^{2n+1} = (V_{\text{re}}^2 + \Delta^2)^n \hat{H}$

for any integer $n$.

---

## ðŸ”¹ Taylor Series Summation

Substituting back into the Taylor series:

\[
\begin{aligned}
\hat{U}(t)
&= \sum_{n=0}^{\infty} \frac{1}{(2n)!} \left( -\frac{i}{\hbar} t \right)^{2n} (V_{\text{re}}^2 + \Delta^2)^n I \\
&\quad + \sum_{n=0}^{\infty} \frac{1}{(2n+1)!} \left( -\frac{i}{\hbar} t \right)^{2n+1} (V_{\text{re}}^2 + \Delta^2)^n \hat{H}
\end{aligned}
\]

Factor terms:

\[
\begin{aligned}
\hat{U}(t)
&= \left[ \sum_{n=0}^{\infty} \frac{1}{(2n)!} \left( -i \frac{\Omega t}{\hbar} \right)^{2n} \right] I \\
&\quad + \left[ \sum_{n=0}^{\infty} \frac{1}{(2n+1)!} \left( -i \frac{\Omega t}{\hbar} \right)^{2n+1} \right] \frac{\hat{H}}{\Omega}
\end{aligned}
\]

where we define the **generalized Rabi frequency**:

$$
\Omega = \sqrt{V_{\text{re}}^2 + \Delta^2}
$$

Recognizing the Taylor series of cosine and sine:

$$
\begin{aligned}
\cos\left( \frac{\Omega t}{\hbar} \right) &= \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} \left( \frac{\Omega t}{\hbar} \right)^{2n} \\
\sin\left( \frac{\Omega t}{\hbar} \right) &= \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n+1)!} \left( \frac{\Omega t}{\hbar} \right)^{2n+1}
\end{aligned}
$$

Thus:

$$
\hat{U}(t) = \cos\left( \frac{\Omega t}{\hbar} \right) I - i \sin\left( \frac{\Omega t}{\hbar} \right) \frac{\hat{H}}{\Omega}
$$


In its final form the propagator becomes:

$$
\hat{U}(t) =
\cos\left( \frac{\Omega t}{\hbar} \right) I
- i \frac{V_{\text{re}} \sigma_x + \Delta \sigma_z}{\Omega} \sin\left( \frac{\Omega t}{\hbar} \right)
$$

written explicitly as a matrix:

$$
\hat{U}(t) =
\begin{pmatrix}
\cos\left( \frac{\Omega t}{\hbar} \right) - i \frac{\Delta}{\Omega} \sin\left( \frac{\Omega t}{\hbar} \right)
&
-i \frac{V_{\text{re}}}{\Omega} \sin\left( \frac{\Omega t}{\hbar} \right) \\
-i \frac{V_{\text{re}}}{\Omega} \sin\left( \frac{\Omega t}{\hbar} \right)
&
\cos\left( \frac{\Omega t}{\hbar} \right) + i \frac{\Delta}{\Omega} \sin\left( \frac{\Omega t}{\hbar} \right)
\end{pmatrix}
$$

## ðŸ”¹ Population Dynamics

We again study the dynamics assuming that the system starts in the  ground state $|\psi(0)\rangle = |g\rangle$ at the initial time $t=0$:
The ground state population as the system evolves can be calculated by applying the time-evolution operator and calculating the absolute square of the population which leads to:

$$
P_g(t) = \cos^2\left( \frac{\Omega t}{\hbar} \right) + \left( \frac{\Delta}{\Omega} \right)^2 \sin^2\left( \frac{\Omega t}{\hbar} \right)
$$

For the excited state we get:

$$
P_e(t) = \left( \frac{V_{\text{re}}}{\Omega} \right)^2 \sin^2\left( \frac{\Omega t}{\hbar} \right)
$$

As it should be the total population is normalized $P_g(t) + P_e(t) = 1$.

---



```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
hbar = 1.0  # Use natural units: hbar = 1
V_re = 1.0  # Coupling strength

# Detuning values
deltas = [0.0, 0.5 * V_re, 1.0 * V_re]
labels = [r'$\Delta = 0$', r'$\Delta = 0.5 V_{\mathrm{re}}$', r'$\Delta = V_{\mathrm{re}}$']
colors = ['blue', 'green', 'red']

# Time grid
t = np.linspace(0, 10 * np.pi, 1000)

# Plot setup
plt.figure(figsize=(9, 6))

for delta, label, color in zip(deltas, labels, colors):
    Omega = np.sqrt(V_re**2 + delta**2)  # Generalized Rabi frequency
    P_e = (V_re / Omega)**2 * np.sin(Omega * t / hbar)**2  # Excited state population
    plt.plot(t, P_e, label=label, color=color, linewidth=2)

# Plot formatting
plt.title('Detuned Rabi Oscillations', fontsize=16)
plt.xlabel('Time $t$', fontsize=14)
plt.ylabel('Excited State Population $P_e(t)$', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)
plt.ylim(-0.05, 1.05)
plt.tight_layout()
plt.show()
```

### ðŸ§  The simulation leads to the following physical insight about the dynamics in the non-resonant case:

- Detuning $\Delta \neq 0$ increases the effective oscillation frequency to $\Omega$.
- The maximum population transfer is reduced to $\left( \frac{V_{\text{re}}}{\Omega} \right)^2$.
- In the limit $\Delta \to 0$, we recover full Rabi oscillations with full transfer between states.

```python
import numpy as np
import matplotlib.pyplot as plt

# Physical constants
hbar = 1.0  # Natural units
V_re = 1.0  # Coupling strength

# Detuning range
Delta = np.linspace(-3 * V_re, 3 * V_re, 500)  # From 0 to 3 * V_re

# Maximum excited state population
P_e_max = V_re**2 / (V_re**2 + Delta**2)

# Plot
plt.figure(figsize=(8,5))
plt.plot(Delta, P_e_max, color='blue', linewidth=2)
plt.title('Maximum Excited State Population vs Detuning', fontsize=16)
plt.xlabel(r'Detuning $\Delta$', fontsize=14)
plt.ylabel(r'Maximum Population $P_e^{\mathrm{max}}$', fontsize=14)
plt.grid(True)
plt.ylim(-0.05, 1.05)
plt.tight_layout()
plt.show()
```

## ðŸ”¸ Spectral Shape of the Maximum Excited State Population

In the previous plot, we observed how the **maximum population transfer** to the excited state $|e\rangle$ depends on the **detuning** $\Delta$.

The functional form is:

$$
P_{e, \mathrm{max}}(\Delta) = \frac{V_{\mathrm{re}}^2}{V_{\mathrm{re}}^2 + \Delta^2}
$$

where $V_{\mathrm{re}}$ is the coupling strength and $\Delta$ is the energy difference between the two levels (detuning). This function is in physics known as the  **Lorentzian function** centered at $\Delta = 0$, with **full width at half maximum (FWHM)** equal to $2 V_{\mathrm{re}}$. Physically, the spectral response is sharp if $V_{\mathrm{re}}$ is small (narrow peak) and broad if $V_{\mathrm{re}}$ is large.

The **Lorentzian spectral profile** is typical for the **resonance behavior** of a two-level quantum system in quantum mechanics (and also for a driven and damped harmonic oscillator in classical mechanics):

- At exact resonance ($\Delta = 0$), the transition probability is maximal ($P_e^{\mathrm{max}} = 1$).
- Away from resonance, the probability **decreases rapidly to zero**.
- The coupling strength $V_{\mathrm{re}}$ determines the "width" of the resonance.

This behavior is encountered in **spectroscopy**, **quantum optics**, and **magnetic resonance** phenomena, where transition probabilities depend sensitively on detuning.
