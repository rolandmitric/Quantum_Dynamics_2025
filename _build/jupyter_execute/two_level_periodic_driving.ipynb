## ðŸ”¸ Dynamics of a Two-Level System Driven by a Periodic Perturbation

We now solve the dynamics of a two-level quantum system driven by a time-periodic external field.

---

## ðŸ”¹ Setup: Time-Dependent Hamiltonian

Consider a two-level system $\{ |g\rangle, |e\rangle \}$ with a Hamiltonian:

$$
\hat{H}(t) = \frac{\hbar}{2} \begin{pmatrix}
0 & \Omega_0 e^{-i\omega t} \\
\Omega_0 e^{i\omega t} & 2\Delta
\end{pmatrix},
$$

where:

- $\Omega_0$ is the **coupling strength** (Rabi frequency),
- $\omega$ is the **driving frequency**,
- $\Delta$ is the **detuning** between the driving field and the system resonance.

The time-dependent SchrÃ¶dinger equation reads:

$$
i\hbar \frac{d}{dt} |\psi(t)\rangle = \hat{H}(t) |\psi(t)\rangle.
$$

We seek an analytic solution.

---

## ðŸ”¹ Step 1: Move to a Rotating Frame

Define a **rotating frame** transformation:

$$
|\psi(t)\rangle = \hat{R}(t) |\tilde{\psi}(t)\rangle
$$

with unitary operator:

$$
\hat{R}(t) = \exp\left( -i \frac{\omega t}{2} \sigma_z \right)
= \begin{pmatrix}
e^{-i\omega t/2} & 0 \\
0 & e^{i\omega t/2}
\end{pmatrix}.
$$

Here, $\sigma_z$ is the Pauli $Z$ matrix.

âœ…  
This transformation **removes** the fast oscillations associated with the driving frequency $\omega$.

---

## ðŸ”¹ Step 2: Derive the Effective SchrÃ¶dinger Equation

Substituting into the SchrÃ¶dinger equation:

\[
i\hbar \frac{d}{dt} \left( \hat{R}(t) |\tilde{\psi}(t)\rangle \right) = \hat{H}(t) \hat{R}(t) |\tilde{\psi}(t)\rangle
\]

Expanding the derivative:

\[
i\hbar \left( \frac{d\hat{R}(t)}{dt} |\tilde{\psi}(t)\rangle + \hat{R}(t) \frac{d|\tilde{\psi}(t)\rangle}{dt} \right) = \hat{H}(t) \hat{R}(t) |\tilde{\psi}(t)\rangle
\]

Multiply from the left by $\hat{R}^\dagger(t)$:

\[
i\hbar \frac{d|\tilde{\psi}(t)\rangle}{dt} = \left( \hat{R}^\dagger(t) \hat{H}(t) \hat{R}(t) - i\hbar \hat{R}^\dagger(t) \frac{d\hat{R}(t)}{dt} \right) |\tilde{\psi}(t)\rangle
\]

Thus, the effective Hamiltonian in the rotating frame is:

$$
\hat{H}_{\text{rot}}(t) = \hat{R}^\dagger(t) \hat{H}(t) \hat{R}(t) - i\hbar \hat{R}^\dagger(t) \frac{d\hat{R}(t)}{dt}.
$$

---

## ðŸ”¹ Step 3: Simplify the Effective Hamiltonian

Compute:

- The derivative:

$$
\frac{d\hat{R}(t)}{dt} = -i\frac{\omega}{2} \sigma_z \hat{R}(t),
$$

thus:

$$
\hat{R}^\dagger(t) \frac{d\hat{R}(t)}{dt} = -i\frac{\omega}{2} \sigma_z.
$$

- The transformation of $\hat{H}(t)$:

Since $\sigma_z$ commutes with itself and rotates $\sigma_x$ and $\sigma_y$, after some algebra you find:

$$
\hat{R}^\dagger(t) \hat{H}(t) \hat{R}(t) = \frac{\hbar}{2} \begin{pmatrix}
0 & \Omega_0 \\
\Omega_0 & 2\Delta
\end{pmatrix}.
$$

âœ…  
Thus, the **effective rotating-frame Hamiltonian** becomes:

$$
\hat{H}_{\text{rot}} = \frac{\hbar}{2} \left( \begin{pmatrix}
0 & \Omega_0 \\
\Omega_0 & 2\Delta
\end{pmatrix}
- \omega \sigma_z \right)
= \frac{\hbar}{2} \begin{pmatrix}
-\omega & \Omega_0 \\
\Omega_0 & 2\Delta + \omega
\end{pmatrix}.
$$

---

## ðŸ”¹ Step 4: Apply the Rotating Wave Approximation (RWA)

In RWA, we assume that the driving frequency $\omega$ is close to resonance, so that $\Delta$ is small compared to $\omega$, and we drop fast-oscillating terms.

This simplifies the detuning to:

$$
\Delta_{\text{eff}} = \Delta + \omega/2,
$$

and the effective Hamiltonian becomes:

$$
\hat{H}_{\text{eff}} = \frac{\hbar}{2} \begin{pmatrix}
0 & \Omega_0 \\
\Omega_0 & 2\Delta
\end{pmatrix},
$$

where now $\Delta$ is interpreted as the **effective detuning** between the drive and the transition.

âœ…  
Now the Hamiltonian is **time-independent**!

---

## ðŸ”¹ Step 5: Solve the Two-Level Dynamics

The time evolution is governed by:

$$
i\hbar \frac{d}{dt} |\tilde{\psi}(t)\rangle = \hat{H}_{\text{eff}} |\tilde{\psi}(t)\rangle
$$

with the initial condition transformed to the rotating frame.

You can solve this exactly:

- The system undergoes **Rabi oscillations** with generalized frequency:

$$
\Omega_R = \sqrt{\Omega_0^2 + \Delta^2}.
$$

- The probability to be in the excited state at time $t$ is:

$$
P_e(t) = \frac{\Omega_0^2}{\Omega_R^2} \sin^2\left( \frac{\Omega_R t}{2} \right).
$$

âœ…  
This describes how the population oscillates between $|g\rangle$ and $|e\rangle$ under the influence of the periodic drive.

---

### ðŸ§  Key Insights

- Moving to the **rotating frame** removes fast oscillations.
- Applying the **Rotating Wave Approximation** simplifies the Hamiltonian.
- **Analytic solutions** exist and describe **Rabi oscillations** driven by the external field.


## ðŸ”¸ Bloch-Siegert Shift: A Correction Beyond the Rotating Wave Approximation

When the rotating wave approximation (RWA) breaks down â€” particularly when the coupling strength $\Omega_0$ is not much smaller than the driving frequency $\omega$ â€” the resonance frequency of the system is **shifted**.

This correction is known as the **Bloch-Siegert shift**.

---

## ðŸ”¹ Setup: Two-Level Hamiltonian with Periodic Driving

The full Hamiltonian (before applying RWA) for a two-level system is:

$$
\hat{H}(t) = \frac{\hbar}{2}
\begin{pmatrix}
0 & \Omega_0 e^{-i\omega t} + \Omega_0 e^{i\omega t} \\
\Omega_0 e^{i\omega t} + \Omega_0 e^{-i\omega t} & 2\Delta
\end{pmatrix}.
$$

The **off-diagonal terms** oscillate at $\pm\omega$ â€” this is why RWA usually drops the fast terms.

---

## ðŸ”¹ Key Observation

- The terms oscillating at $+2\omega$ (the counter-rotating terms) usually average out if $\omega$ is large.
- But if $\Omega_0$ is not very small compared to $\omega$, the counter-rotating terms **produce a small but systematic effect**:  
  they **shift the effective resonance frequency**.

âœ… This shift is what we calculate next.

---

## ðŸ”¹ Step 1: Identify the Fast and Slow Terms

Expand the driving field:

$$
\Omega_0 e^{-i\omega t} + \Omega_0 e^{i\omega t} = 2\Omega_0 \cos(\omega t).
$$

Thus, the Hamiltonian becomes:

$$
\hat{H}(t) = \frac{\hbar}{2}
\left(
\Omega_0 \cos(\omega t) \, \sigma_x + \Delta \, \sigma_z
\right).
$$

---

## ðŸ”¹ Step 2: Go to the Rotating Frame

We perform a unitary transformation:

$$
\hat{R}(t) = \exp\left( -i\frac{\omega t}{2} \sigma_z \right).
$$

The effective Hamiltonian in the rotating frame becomes:

$$
\hat{H}_{\text{rot}}(t) = \frac{\hbar}{2}
\left[
\Delta \sigma_z + \Omega_0 \left( \cos(2\omega t) \sigma_x + \sin(2\omega t) \sigma_y \right)
+ \Omega_0 \sigma_x
\right]
- \frac{\hbar \omega}{2} \sigma_z.
$$

Simplifying:

$$
\hat{H}_{\text{rot}}(t) = \frac{\hbar}{2}
\left[
(\Delta - \omega) \sigma_z + \Omega_0 \sigma_x
+ \Omega_0 \left( \cos(2\omega t) \sigma_x + \sin(2\omega t) \sigma_y \right)
\right].
$$

âœ…  
Thus, even in the rotating frame, there are **fast oscillating terms at $2\omega$**.

---

## ðŸ”¹ Step 3: Treat Fast Oscillations Perturbatively

Now:

- $\cos(2\omega t)$ and $\sin(2\omega t)$ oscillate very fast compared to the timescale of $\Omega_0$.
- Thus, you can **treat the fast terms perturbatively**.

The correction to the energy levels due to the fast terms can be estimated via **second-order perturbation theory**.

Result:

- The counter-rotating terms cause a shift in the effective energy splitting.
- This shift in the resonance frequency is called the **Bloch-Siegert shift** $\delta \omega_{\text{BS}}$.

Explicitly, the shift is:

$$
\boxed{
\delta \omega_{\text{BS}} = \frac{\Omega_0^2}{4\omega}
}.
$$

âœ…  
Thus, the true resonance occurs not when $\Delta = 0$, but when:

$$
\Delta_{\text{resonance}} = \delta \omega_{\text{BS}}.
$$

---

## ðŸ”¹ Step 4: Physical Interpretation

- The Bloch-Siegert shift **pushes** the effective resonance frequency slightly **higher**.
- It becomes important when $\Omega_0$ is not much smaller than $\omega$.

âœ…  
Thus, **RWA fails** to predict the exact resonance condition when the drive strength $\Omega_0$ is **not negligible compared to** $\omega$.

---

### ðŸ§  Summary

- **Counter-rotating terms** oscillating at $2\omega$ cause a small systematic **shift** in the resonance condition.
- The Bloch-Siegert shift is:

$$
\delta \omega_{\text{BS}} = \frac{\Omega_0^2}{4\omega}.
$$

- **When $\Omega_0/\omega$ is not very small**, this shift must be taken into account.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Omega_0 = 0.2        # Rabi frequency (in units where hbar=1)
omega_drive = 1.0    # Driving frequency
Delta = 0.0          # Detuning (initially assume exact resonance)

# Derived quantities
delta_BS = Omega_0**2 / (4 * omega_drive)  # Bloch-Siegert shift

# Time array
t = np.linspace(0, 100, 1000)

# Effective Rabi frequencies
Omega_R_RWA = np.sqrt(Omega_0**2 + Delta**2)                     # RWA
Omega_R_BS = np.sqrt(Omega_0**2 + (Delta + delta_BS)**2)          # Corrected

# Populations
P_e_RWA = (Omega_0**2 / Omega_R_RWA**2) * np.sin(Omega_R_RWA * t / 2)**2
P_e_BS = (Omega_0**2 / Omega_R_BS**2) * np.sin(Omega_R_BS * t / 2)**2

# Plot
plt.figure(figsize=(10,6))
plt.plot(t, P_e_RWA, label="RWA", linestyle='--')
plt.plot(t, P_e_BS, label="Including Bloch-Siegert shift", linestyle='-')
plt.title("Excited State Population vs Time")
plt.xlabel(r"Time $t$")
plt.ylabel(r"Population $P_e(t)$")
plt.legend()
plt.grid(True)
plt.show()
```
