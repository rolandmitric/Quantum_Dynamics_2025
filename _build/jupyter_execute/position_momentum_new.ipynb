## Construction of Position and Momentum Observables via Translation Symmetry

In this section, we introduce the observables corresponding to position and momentum in a semi-rigorous manner, using physical assumptions and basic operator theory. We avoid referring to "position eigenstates," which do not exist in the Hilbert space of normalisable wavefunctions.

### 1. Postulating the Position Operator

We assume that there exists a self-adjoint operator $\hat{x}$, called the **position operator**, with continuous spectrum, defined on a dense domain in the Hilbert space $\mathcal{H} = L^2(\mathbb{R})$. We consider a normalisable state $|\psi\rangle \in \mathcal{H}$ such that:

$$
\langle \psi | \hat{x} | \psi \rangle = x_0
$$

This expectation value corresponds to the mean position of a particle in the state $|\psi\rangle$.

---

### 2. Defining the Translation Operator

We now postulate the existence of a **unitary translation operator** $T(a)$ which translates the expectation value of position by a real displacement $a$:

$$
\langle \psi | T(a)^\dagger \hat{x} T(a) | \psi \rangle = x_0 + a
$$

This implies that $T(a)$ implements a spatial shift. The action of $T(a)$ on the position operator satisfies the operator identity:

$$
T(a)^\dagger \hat{x} T(a) = \hat{x} + a I
$$

This identity can be verified for all states in a dense subset of the Hilbert space and is consistent with translation symmetry in quantum mechanics.

---

### 3. Momentum as the Generator of Translations

By **Stone's theorem**, a strongly continuous one-parameter group of unitary operators $T(a)$ must be generated by a self-adjoint operator. Therefore, there exists a self-adjoint operator $\hat{p}$ such that:

$$
T(a) = e^{-i a \hat{p} / \hbar}
$$

This defines the **momentum operator** $\hat{p}$ as the generator of translations.

---

### 4. Deriving the Canonical Commutation Relation

We now derive the fundamental commutation relation between position and momentum.

Differentiate the identity

$$
T(a)^\dagger \hat{x} T(a) = \hat{x} + a I
$$

with respect to $a$ and evaluate at $a = 0$:

$$
\frac{d}{da} \left( e^{i a \hat{p}/\hbar} \hat{x} e^{-i a \hat{p}/\hbar} \right)\Big|_{a=0}
= \frac{i}{\hbar} [\hat{p}, \hat{x}] = I
$$

This yields the **canonical commutation relation**:

$$
[\hat{x}, \hat{p}] = i \hbar
$$

---

### 5. Representation in Position Space

Once the commutation relation is established, one can realize the operators concretely in the Hilbert space $\mathcal{H} = L^2(\mathbb{R})$ as:

$$
\hat{x} \psi(x) = x \psi(x), \quad
\hat{p} \psi(x) = -i \hbar \frac{d}{dx} \psi(x)
$$

These forms are consistent with the commutation relation and with the interpretation of $\hat{p}$ as the generator of spatial translations.

---

### Summary

- The position operator $\hat{x}$ is postulated as a self-adjoint operator with continuous spectrum.
- A unitary translation operator $T(a)$ is introduced such that it shifts the expectation value of position.
- The momentum operator $\hat{p}$ is defined via Stoneâ€™s theorem as the generator of $T(a)$.
- The canonical commutation relation $[\hat{x}, \hat{p}] = i \hbar$ is derived without invoking ill-defined position eigenstates.
