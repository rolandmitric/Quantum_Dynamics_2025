# Two level system driven by a harmonic field 

We now study the dynamics of a two-level system under a periodic driving potential of the form  

$$
V(t) = V_0 \cdot \cos(\omega t),
$$  

where $\omega$ is the driving frequency.  

The Hamiltonian reads:

$$
H = \frac{1}{2} \Delta \sigma_z + V_0 \cos(\omega t) \, \sigma_x.
$$

In order to make the problem analytically tractable, we introduce an approximation that is commonly known in the literature as the **rotating wave approximation (RWA)**.  

The idea is as follows:  
By expanding the cosine term,

$$
\cos(\omega t) = \frac{1}{2} \left(e^{i \omega t} + e^{-i \omega t} \right) = \frac{1}{2} e^{i \omega t} \left(1 + e^{-2 i \omega t} \right),
$$  

the Hamiltonian can be written (in matrix form) as:

$$
V(t) = \begin{bmatrix}
0 & \frac{1}{2} V_0 e^{i \omega t} \left(1 + e^{-2 i \omega t} \right) \\
\frac{1}{2} V_0 e^{-i \omega t} \left(1 + e^{2 i \omega t} \right) & 0
\end{bmatrix}.
$$

Notice that in the second row we have extracted the term $e^{-i \omega t}$ such that the leading terms are Hermitian conjugates of each other.  
We see that the additional terms ($e^{-2 i \omega t}$ and $e^{2 i \omega t}$) rotate with double the frequency and in the opposite direction compared to the $e^{\pm i \omega t}$ terms.

We will show below that these rapidly oscillating terms can be neglected if the driving frequency is close to resonance and the coupling term is not too large. This will be justified by comparison with the numerically exact solution later.  

This is the essence of the **rotating wave approximation**, often used in the literature:

$$
H = \frac{1}{2} \Delta \sigma_z + V_0 \cos(\omega t) \, \sigma_x
\quad \Rightarrow \quad
H_{\text{RWA}} = \frac{1}{2} \Delta \sigma_z + \frac{1}{2} V_0 e^{i \omega t} \sigma_+ + \frac{1}{2} V_0 e^{-i \omega t} \sigma_-.
$$

Using the relations:

$$
\begin{aligned}
\sigma_+ &= \sigma_x + i \sigma_y, \\
\sigma_- &= \sigma_x - i \sigma_y,
\end{aligned}
$$  

we express the Hamiltonian in terms of the raising and lowering operators.


We may write:
$$
\begin{aligned}
H_{\text{RWA}} 
&= \frac{1}{2} \Delta \sigma_z + \frac{1}{2} V_0 e^{i \omega t} (\sigma_x + i \sigma_y) + \frac{1}{2} V_0 e^{-i \omega t} (\sigma_x - i \sigma_y) \\
&= \frac{1}{2} \Delta \sigma_z + V_0 \cos(\omega t) \, \sigma_x - V_0 \sin(\omega t) \, \sigma_y
\end{aligned}
$$

We now proceed to calculate the time-evolution operator in the RWA approximation:
$$
\frac{dU}{dt} = H_{\text{RWA}} U 
= \left[ \frac{1}{2} \Delta \sigma_z + V_0 \left( \cos(\omega t) \, \sigma_x - \sin(\omega t) \, \sigma_y \right) \right] U
$$

To simplify the solution, we move to a **rotating frame of reference**.  
The intuition behind this is that in the rotating frame the coupling term appears static, reducing the problem to a previously solved case of a constant perturbation.  
Thus, we define:
$$
U(t) = e^{-\frac{i}{2} \omega t \sigma_z} \, \tilde{U}(t)
$$

The time derivative of $U$ becomes:
$$
\frac{dU}{dt} = -\frac{i}{2} \omega \sigma_z U + e^{-\frac{i}{2} \omega t \sigma_z} \frac{d\tilde{U}}{dt}
$$

Inserting into the time-dependent Schrödinger equation for $U(t)$ gives:
$$
\frac{\hbar \omega}{2} \sigma_z e^{-\frac{i}{2} \omega t \sigma_z} \tilde{U}
+ i \hbar e^{-\frac{i}{2} \omega t \sigma_z} \frac{d\tilde{U}}{dt}
= \frac{1}{2} \Delta \sigma_z e^{-\frac{i}{2} \omega t \sigma_z} \tilde{U}
+ \left( V_0 \cos(\omega t) \, \sigma_x - V_0 \sin(\omega t) \, \sigma_y \right) e^{-\frac{i}{2} \omega t \sigma_z} \tilde{U}
$$

By defining $\Delta = \hbar \omega_0$, we can write:
$$
i \hbar e^{-\frac{i}{2} \omega t \sigma_z} \frac{d\tilde{U}}{dt}
= \frac{1}{2} \hbar (\omega_0 - \omega) \sigma_z e^{-\frac{i}{2} \omega t \sigma_z} \tilde{U}
+ \left( V_0 \cos(\omega t) \, \sigma_x - V_0 \sin(\omega t) \, \sigma_y \right) e^{-\frac{i}{2} \omega t \sigma_z} \tilde{U}
$$

We can now multiply both sides by $e^{\frac{i}{2} \omega t \sigma_z}$ from the left, getting:


We now have:
$$
\frac{d\tilde{U}}{dt} = \left[ \frac{1}{2} \hbar(\omega_0 - \omega) \sigma_z + V_0 \, e^{\frac{i}{2} \omega t \sigma_z} \left( \cos(\omega t) \, \sigma_x - \sin(\omega t) \, \sigma_y \right) e^{-\frac{i}{2} \omega t \sigma_z} \right] \tilde{U}
$$

To simplify this, we need to compute the similarity transformations of the Pauli matrices under rotations generated by $\sigma_z$. Specifically:

### Rotation of Pauli matrices
We use the identity
$$
e^{i \theta \sigma_z} \, \sigma_x \, e^{-i \theta \sigma_z} = \cos(2\theta) \, \sigma_x + \sin(2\theta) \, \sigma_y,
$$
and similarly,
$$
e^{i \theta \sigma_z} \, \sigma_y \, e^{-i \theta \sigma_z} = \cos(2\theta) \, \sigma_y - \sin(2\theta) \, \sigma_x.
$$

These follow from the **Baker-Campbell-Hausdorff** formula or by applying the **Euler operator identity**:
$$
e^{i A x} = \cos(x) \cdot I + i A \sin(x),
$$
which is valid when $A^2 = I$. For Pauli matrices, $\sigma_z^2 = I$.

### Apply to the time-dependent rotation
Set $\theta = \frac{\omega t}{2}$. Then:
$$
\begin{aligned}
e^{i \frac{\omega t}{2} \sigma_z} \sigma_x e^{-i \frac{\omega t}{2} \sigma_z} &= \cos(\omega t) \, \sigma_x + \sin(\omega t) \, \sigma_y, \\
e^{i \frac{\omega t}{2} \sigma_z} \sigma_y e^{-i \frac{\omega t}{2} \sigma_z} &= \cos(\omega t) \, \sigma_y - \sin(\omega t) \, \sigma_x.
\end{aligned}
$$

Therefore, the rotated expression becomes:
$$
\begin{aligned}
& e^{i \frac{\omega t}{2} \sigma_z} \left( \cos(\omega t) \, \sigma_x - \sin(\omega t) \, \sigma_y \right) e^{-i \frac{\omega t}{2} \sigma_z} \\
&= \cos(\omega t) \left[ \cos(\omega t) \, \sigma_x + \sin(\omega t) \, \sigma_y \right] 
- \sin(\omega t) \left[ \cos(\omega t) \, \sigma_y - \sin(\omega t) \, \sigma_x \right] \\
&= \left( \cos^2(\omega t) + \sin^2(\omega t) \right) \sigma_x + 2 \cos(\omega t) \sin(\omega t) \sigma_y - 2 \cos(\omega t) \sin(\omega t) \sigma_y \\
&= \sigma_x.
\end{aligned}
$$

Hence, the entire rotated term simplifies beautifully to:
$$
e^{i \frac{\omega t}{2} \sigma_z} \left( \cos(\omega t) \, \sigma_x - \sin(\omega t) \, \sigma_y \right) e^{-i \frac{\omega t}{2} \sigma_z} = \sigma_x.
$$

### Final simplified equation
The time-evolution equation in the rotating frame now reads:
$$
\frac{d\tilde{U}}{dt} = \left[ \frac{1}{2} \hbar(\omega_0 - \omega) \sigma_z + V_0 \sigma_x \right] \tilde{U}.
$$

This is a **time-independent** Schrödinger equation in the rotating frame.  
It describes a two-level system under a static effective Hamiltonian:
$$
H_{\text{eff}} = \frac{1}{2} \hbar(\omega_0 - \omega) \sigma_z + V_0 \sigma_x.
$$


## Solution Using Time-Evolution Operator

In order to propagate the system in the rotating frame we can now use the time-evolution operator that we constructed for a non-resonant coupling case. We get for $U(t)$:

$$
U(t) = \cos\left(\frac{\Omega_R t}{2}\right) \, \mathbb{I}
- i \sin\left(\frac{\Omega_R t}{2}\right)
\left( \frac{2V_0}{\hbar \Omega_R} \sigma_x + \frac{\delta}{\Omega_R} \sigma_z \right).
$$

---

Assume the initial state is the ground state:

$$
|\psi(0)\rangle = |0\rangle = \begin{pmatrix}1 \\ 0\end{pmatrix}.
$$

Then the evolved state is:
$$
|\psi(t)\rangle = U(t) |\psi(0)\rangle.
$$

Carrying out the multiplication:
$$
|\psi(t)\rangle =
\left[
\cos\left( \frac{\Omega_R t}{2} \right)
- i \frac{\delta}{\Omega_R} \sin\left( \frac{\Omega_R t}{2} \right)
\right] \begin{pmatrix}1 \\ 0\end{pmatrix}
- i \frac{2V_0}{\hbar \Omega_R} \sin\left( \frac{\Omega_R t}{2} \right) \begin{pmatrix}0 \\ 1\end{pmatrix}.
$$

So the wavefunction becomes:
$$
|\psi(t)\rangle =
\begin{pmatrix}
\cos\left( \frac{\Omega_R t}{2} \right)
- i \frac{\delta}{\Omega_R} \sin\left( \frac{\Omega_R t}{2} \right) \\
- i \frac{2V_0}{\hbar \Omega_R} \sin\left( \frac{\Omega_R t}{2} \right)
\end{pmatrix}.
$$

---


From this, the population in the **excited state** is:

$$
P_{\text{excited}}(t) =
\left| - i \frac{2V_0}{\hbar \Omega_R} \sin\left( \frac{\Omega_R t}{2} \right) \right|^2
= \frac{4V_0^2}{\hbar^2 \Omega_R^2} \sin^2\left( \frac{\Omega_R t}{2} \right).
$$

And since $\Omega_R^2 = \delta^2 + \left( \frac{2V_0}{\hbar} \right)^2$, we recover the familiar expression:

$$
P_{\text{excited}}(t) =
\frac{\left(\frac{2V_0}{\hbar}\right)^2}
{\delta^2 + \left(\frac{2V_0}{\hbar}\right)^2}
\cdot \sin^2\left( \frac{\Omega_R t}{2} \right).
$$

---

This completes the derivation using the general time-evolution operator for a two-level system with constant coupling.


# The `MultiLevel` Class Implementation
 
 We implement a simple Python class to define and propagate quantum states in an N-level system.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
from scipy.linalg import expm

class MultiLevel:
    def __init__(self, ndim=2):
        self.ndim = ndim
        self.h = np.zeros((ndim, ndim), dtype=complex)  
        self.V = lambda t: np.zeros((ndim, ndim), dtype=complex)  
        
    def set_hamiltonian(self, diagonal):
        np.fill_diagonal(self.h, diagonal)

    def set_coupling(self, V_func):
        self.V = V_func

    def getH(self, t):
        return self.h + self.V(t)

    def propagate(self, psi0, dt, nstep):

        psi0 = np.asarray(psi0, dtype=complex)
        psi_t = np.zeros((nstep + 1, self.ndim), dtype=complex)
        psi_t[0] = psi0

        for i in range(1, nstep + 1):
            t = (i - 1) * dt
            H = self.getH(t)
            U = expm(-1j * H * dt)
            psi_t[i] = U @ psi_t[i - 1]

        return psi_t
```

#  Two-Level System with harmonic coupling: Numerical propagation

 Consider a two-level system with energies $ E_0 = 0 $, $ E_1 = \Delta $, and a Gaussian coupling:

 $$
 V_{01} = V_{10} = V_0 \cos(\omega t)
 $$

```python
Delta = 1.0     # Energy difference
V0 = 0.1        # Coupling strength
omega = 1.0     # Frequency of a light-field
coupling = lambda t: np.array([[0, V0*np.cos(omega*t) ],
                               [V0 *np.cos(omega*t) , 0]], dtype=complex)
```

```python
sys = MultiLevel(ndim=2)
sys.set_hamiltonian([0, Delta])
sys.set_coupling(coupling)
psi0=np.array([1.0,0.0],dtype = complex)
```

```python
nstep = 1000
dt = 0.1
psi_t = sys.propagate(psi0, dt, nstep)
```

```python
pop = np.abs(psi_t)**2
coh = np.einsum('ti,tj->tij',psi_t.conj(),psi_t)
time = np.linspace(0,100.0,1000+1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(time, pop[:,0], lw = 3.0, color = "red")
plt.plot(time, pop[:,1], lw = 3.0, color = "blue")
plt.plot(time, coh[:,0,1], "--")
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Two-Level System Driven by Gaussian Coupling')
plt.grid(True)
```

## Comparison between the RWA and numerical propagation for a resonant coupling

```python
# Parameters
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

Δ = 1.0       # Energy splitting
V = .3    # Driving strength
ω = 1.0      # Driving frequency (resonant)
dt = 0.01
t_max = 1000
times = np.arange(0, t_max, dt)

# Initial state: ground state |g> = [1, 0]
ψ = np.array([1.0, 0.0], dtype=complex)
ψs = []

# Time evolution under full Hamiltonian
for t in times:
    H = (Δ/2)*np.array([[1,0],[0,-1]]) + V*np.cos(ω*t)*np.array([[0,1],[1,0]])
    U = expm(-1j * H * dt)
    ψ = U @ ψ
    ψs.append(np.abs(ψ[1])**2)  # Excited state population

# RWA valid for resonance: ω ≈ Δ
# Analytic RWA prediction: Rabi flopping with freq V/2
P_rwa = np.sin(0.5 * V * times)**2


# Plot comparison
plt.plot(times, ψs, color = "red", label='Numerical', lw=2)
plt.plot(times, P_rwa, color = "blue", label='RWA', lw=2)
plt.xlabel('Time')
plt.ylabel('Excited State Population')
plt.title('Numerical vs RWA')
plt.legend()
plt.grid(True)
plt.xlim(900,1000)
plt.show()
```
