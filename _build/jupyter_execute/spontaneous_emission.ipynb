# üßÆ Goal

Derive the **spontaneous decay rate** $\Gamma$ of an excited atomic state $|e\rangle$, coupled to the vacuum electromagnetic field, **without invoking Fermi‚Äôs golden rule directly**, but using **time-dependent perturbation theory** and a **rigorous integral analysis**.

---

## 1. Total Hamiltonian and Initial State

The Hilbert space is $\mathcal{H}_{\text{atom}} \otimes \mathcal{H}_{\text{field}}$. The total Hamiltonian is:

$$
\hat{H} = \hat{H}_0 + \hat{V}
$$

with

* $\hat{H}_0 = \hat{H}_\text{atom} + \hat{H}_\text{field}$
* $\hat{V} = -\hat{\mathbf{d}} \cdot \hat{\mathbf{E}}(\mathbf{r})$

We expand the total wavefunction in the following ansatz:

$$
|\Psi(t)\rangle = c_e(t) |e\rangle \otimes |0\rangle + \sum_{\mathbf{k},\lambda} c_{g,\mathbf{k},\lambda}(t) |g\rangle \otimes |\mathbf{k},\lambda\rangle
$$

We assume the atom starts in the excited state with no photons:

$$
|\Psi(0)\rangle = |e\rangle \otimes |0\rangle \quad \Rightarrow \quad c_e(0) = 1, \quad c_{g,\mathbf{k},\lambda}(0) = 0
$$

---

## 2. Schr√∂dinger Equation in the Interaction Picture

We work in the **interaction picture**, where operators evolve under $\hat{H}_0$, and the state evolves under:

$$
i\hbar \frac{d}{dt} |\Psi_I(t)\rangle = \hat{V}_I(t) |\Psi_I(t)\rangle
$$

with

$$
\hat{V}_I(t) = e^{i\hat{H}_0 t/\hbar} \hat{V} e^{-i\hat{H}_0 t/\hbar}
$$

Inserting the ansatz, we get the following system of equations:

$$
\begin{aligned}
\frac{d}{dt} c_e(t) &= -\frac{i}{\hbar} \sum_{\mathbf{k},\lambda} \langle e, 0 | \hat{V}_I(t) | g, \mathbf{k},\lambda \rangle c_{g,\mathbf{k},\lambda}(t) \\
\frac{d}{dt} c_{g,\mathbf{k},\lambda}(t) &= -\frac{i}{\hbar} \langle g, \mathbf{k},\lambda | \hat{V}_I(t) | e, 0 \rangle c_e(t)
\end{aligned}
$$

---

## 3. First-Order Equation for the Photon Modes

From the second equation:

$$
\frac{d}{dt} c_{g,\mathbf{k},\lambda}(t) = -\frac{i}{\hbar} \langle g | \hat{\mathbf{d}} | e \rangle \cdot \langle \mathbf{k},\lambda | \hat{\mathbf{E}}_I(\mathbf{r},t) | 0 \rangle c_e(t)
$$

The quantized electric field in the interaction picture:

$$
\hat{\mathbf{E}}_I(\mathbf{r},t) = i \sum_{\mathbf{k},\lambda} \sqrt{\frac{\hbar \omega_k}{2\varepsilon_0 V}} \mathbf{e}_{\mathbf{k},\lambda} \hat{a}_{\mathbf{k},\lambda} e^{i(\mathbf{k}\cdot\mathbf{r} - \omega_k t)} + \text{H.c.}
$$

Only the annihilation operator $\hat{a}_{\mathbf{k},\lambda}$ contributes:

$$
\langle \mathbf{k},\lambda | \hat{\mathbf{E}}_I(\mathbf{r},t) | 0 \rangle = i \sqrt{\frac{\hbar \omega_k}{2\varepsilon_0 V}} \mathbf{e}_{\mathbf{k},\lambda} e^{i(\mathbf{k}\cdot\mathbf{r} - \omega_k t)}
$$

Thus:

$$
\frac{d}{dt} c_{g,\mathbf{k},\lambda}(t) = -i g_{\mathbf{k},\lambda} e^{-i\omega_k t} c_e(t)
$$

where

$$
g_{\mathbf{k},\lambda} := \frac{1}{\hbar} \sqrt{\frac{\hbar \omega_k}{2\varepsilon_0 V}} \left( \mathbf{e}_{\mathbf{k},\lambda} \cdot \mathbf{d}_{eg} \right) e^{i\mathbf{k}\cdot\mathbf{r}}
$$

Integrate:

$$
c_{g,\mathbf{k},\lambda}(t) = -i g_{\mathbf{k},\lambda} \int_0^t e^{-i\omega_k t'} c_e(t') dt'
$$

---

## 4. Integro-Differential Equation for $c_e(t)$

Insert this into the first equation:

$$
\frac{d}{dt} c_e(t) = -\sum_{\mathbf{k},\lambda} |g_{\mathbf{k},\lambda}|^2 \int_0^t e^{i\omega_k (t - t')} c_e(t') dt'
$$

We now define the **memory kernel**:

$$
K(t - t') = \sum_{\mathbf{k},\lambda} |g_{\mathbf{k},\lambda}|^2 e^{i\omega_k (t - t')}
$$

So the equation becomes:

$$
\frac{d}{dt} c_e(t) = -\int_0^t K(t - t') c_e(t') dt'
$$

---

## 5. Continuum Limit and Density of States

We now let the mode sum become an integral:

$$
\sum_{\mathbf{k},\lambda} \rightarrow \sum_{\lambda=1,2} \int \frac{V d^3 \mathbf{k}}{(2\pi)^3}
$$

$$
|g_{\mathbf{k},\lambda}|^2 = \frac{\omega_k}{2\hbar\varepsilon_0 V} |\mathbf{e}_{\mathbf{k},\lambda} \cdot \mathbf{d}_{eg}|^2
$$

Then:

$$
K(t - t') = \int_0^\infty d\omega \, J(\omega) e^{i\omega (t - t')}
$$

with the **spectral density**:

$$
J(\omega) = \frac{\omega^3}{3\pi^2 \hbar \varepsilon_0 c^3} |\mathbf{d}_{eg}|^2
$$

---

## 6. Markov Approximation and Exponential Decay

The integral equation:

$$
\frac{d}{dt} c_e(t) = -\int_0^t K(t - t') c_e(t') dt'
$$

is a **Volterra equation of the second kind**. In the **Markovian limit** (long times $t \gg 1/\omega$), we approximate:

$$
c_e(t') \approx c_e(t)
$$

and extend the upper limit to infinity (since the kernel is rapidly oscillating):

$$
\frac{d}{dt} c_e(t) \approx -c_e(t) \int_0^\infty K(\tau) d\tau
$$

This gives:

$$
\frac{d}{dt} c_e(t) = -\Gamma c_e(t)
$$

with

$$
\Gamma = \int_0^\infty K(\tau) d\tau = \int_0^\infty d\omega \, J(\omega) \int_0^\infty e^{i\omega \tau} d\tau
$$

The last integral is evaluated using the **Plemelj formula**:

$$
\int_0^\infty e^{i\omega \tau} d\tau = \pi \delta(\omega) + i \mathcal{P} \left( \frac{1}{\omega} \right)
$$

Only the $\delta(\omega - \omega_0)$ contributes at the transition frequency $\omega_0$. So:

$$
\Gamma = 2\pi J(\omega_0) = \frac{\omega_0^3}{3\pi \varepsilon_0 \hbar c^3} |\mathbf{d}_{eg}|^2
$$

---

## ‚úÖ Final Result

$$
\boxed{\Gamma = \frac{\omega_0^3}{3\pi \varepsilon_0 \hbar c^3} |\mathbf{d}_{eg}|^2}
$$

---

## ‚úîÔ∏è Remarks

* This derivation **does not invoke Fermi's Golden Rule** but instead treats the system **as a Volterra integral equation** derived from the Schr√∂dinger equation.
* The **Markov approximation** and **Plemelj identity** play key roles in extracting the exponential decay.
* The derivation is **nonperturbative** in form but uses a **weak coupling limit** (Born-Markov) to obtain exponential decay.
