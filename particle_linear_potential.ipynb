## Particle in a Linear Potential


As an application of the position and momentum representation we now consider a particle moving in a linear potential of the form: 

$$
V(x) = F \cdot x.
$$

We wish to solve the stationary Schr√∂dinger equation

$$
\hat{H}|\psi\rangle = E |\psi\rangle,
$$  

with the Hamiltonian given by

$$
\hat{H} = \frac{\hat{p}^2}{2m} + F \cdot \hat{x}.
$$

Projecting onto the position eigenstate $|x\rangle$, we obtain:

$$
\frac{1}{2m} \langle x| \hat{p}^2 |\psi\rangle + F \langle x|\hat{x}|\psi\rangle = E \langle x|\psi\rangle.
$$

Using $\hat{p} = -i\hbar \frac{d}{dx}$, this becomes:

$$
-\frac{\hbar^2}{2m} \frac{d^2\psi(x)}{dx^2} + F x\, \psi(x) = E\, \psi(x).
$$

This is a second-order differential equation with a non-constant potential. In the position representation this is a very hard problem since the solution of this equation cannot be represented by elementary functions that we like so much.


### Momentum Representation

We now solve the equation in the **momentum representation**, where

$$
\hat{x} = i\hbar \frac{d}{dp}.
$$

Projecting onto $\langle p|$, we get:

$$
\langle p| \frac{\hat{p}^2}{2m} |\psi\rangle + F \langle p|\hat{x}|\psi\rangle = E \langle p|\psi\rangle,
$$

which becomes:

$$
\frac{p^2}{2m} \phi(p) - F \cdot i\hbar \frac{d\phi(p)}{dp} = E \phi(p),
$$

or rearranged:

$$
F (i\hbar) \frac{d\phi}{dp} = \left( \frac{p^2}{2m} - E \right) \phi(p),
$$
and thus:

$$
\frac{d\phi}{dp} = \frac{i}{\hbar} \cdot \frac{1}{2m F} \left( p^2 - 2mE \right) \phi(p).
$$


This is a first-order differential equation with separated variables that can be easily solved:

$$
\frac{d\phi}{\phi} = \frac{i}{\hbar} \cdot \frac{1}{2m F} \left( p^2 - 2mE \right) dp.
$$

Notice that while transforming to the momentum representation we got first order equation from a second one.
Integration gives:

$$
\ln \phi(p) = \frac{i}{\hbar} \cdot \frac{1}{2mF} \left( \frac{p^3}{3} - 2mE\, p \right) + C,
$$

so that

$$
\phi(p) = C \, \exp \left[ \frac{i}{\hbar} \cdot \frac{1}{2mF} \left( \frac{p^3}{3} - 2mE\, p \right) \right].
$$


We now transform back to the position representation via Fourier transform:

$$
\psi(x) = \frac{1}{\sqrt{2\pi\hbar}} \int_{-\infty}^{+\infty} dp \, e^{\frac{i}{\hbar}px} \, \phi(p),
$$

which becomes:

$$
\psi(x) = \frac{C}{\sqrt{2\pi\hbar}} \int_{-\infty}^{+\infty} dp \, 
\exp\left[ \frac{i}{\hbar} \left( p x + \frac{1}{2mF} \left( \frac{p^3}{3} - 2mE\,p \right) \right) \right].
$$

Although this integral cannot be evaluated in closed form using elementary functions, it defines the **exact solution** to the problem. The result is closely related to the Airy function solution obtained in position space. We will in the next section develop an asymptotic form of the solution by using an important technique involving stationary phase integration.
